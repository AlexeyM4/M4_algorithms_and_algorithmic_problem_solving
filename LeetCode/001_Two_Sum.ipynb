# Задача: Сумма двух чисел *(Two Sum)*

## Условие  
Дан массив целых чисел `nums` и целевое значение `target`.  
Требуется определить индексы двух элементов массива, сумма которых равна значению `target`.

Предполагается, что для каждого входного набора данных существует ровно одно решение.  
Использование одного и того же элемента более одного раза не допускается.  
Порядок возвращаемых индексов не имеет значения.

---

## Примеры

**Пример 1**  
Входные данные:  
`nums = [2, 7, 11, 15]`, `target = 9`  
Выходные данные:  
`[0, 1]`  
Пояснение: `nums[0] + nums[1] = 2 + 7 = 9`

**Пример 2**  
Входные данные:  
`nums = [3, 2, 4]`, `target = 6`  
Выходные данные:  
`[1, 2]`

**Пример 3**  
Входные данные:  
`nums = [3, 3]`, `target = 6`  
Выходные данные:  
`[0, 1]`

---

## Ограничения
- `2 <= nums.length <= 10⁴` — `nums.length` обозначает **количество элементов массива** 
- `-10⁹ <= nums[i] <= 10⁹`  
- `-10⁹ <= target <= 10⁹`  
- Существует только одно корректное решение.

---

## Дополнительное требование
Необходимо предложить алгоритм с временной сложностью меньше `O(n²)`.




## Решение 1: Наивный перебор пар элементов

В данном подходе реализован **полный перебор всех возможных пар элементов массива** для поиска суммы, равной целевому значению `target`.  

### Алгоритм
1. Для каждого элемента массива `nums[i]` вычисляется разность `r = target - nums[i]`.  
2. Выполняется вложенный цикл по всем элементам с индексами `j > i`.  
3. Если находится элемент `nums[j]`, равный `r`, возвращаются индексы `[i, j]`.  
4. Если ни одна пара не удовлетворяет условию, возвращается пустой список.

### Особенности
- Гарантируется использование **разных индексов**, что соответствует ограничению задачи о недопустимости повторного использования одного и того же элемента.  
- Временная сложность данного решения составляет `O(n²)`, где `n` — количество элементов массива.  
- Простота реализации делает метод наглядным, однако для больших массивов он является неэффективным.

### Вывод
Метод является прямолинейным и проверяет все возможные пары, что гарантирует корректность при малых и средних размерах массива, но не удовлетворяет требованию улучшенной временной сложности `O(n²)`.

```python
def two_sum_1(nums: list[int], target: int) -> list[int]:
    for i in range(len(nums) - 1):
        r = target - nums[i]
        for j in range(i + 1, len(nums)):
            if nums[j] == r:
                return [i, j]     
    return []


# Тестирование 
assert two_sum_1([1, 3, 4], 4) == [0, 1]
assert two_sum_1([2, 7, 11, 15], 9) == [0,1]
assert two_sum_1([3, 2, 4], 6) == [1,2]
assert two_sum_1([3, 3], 6) == [0,1]
assert two_sum_1([-1, -2, -3, -4, -5], -8) == [2,4]
assert two_sum_1([-3, 4, 3, 90], 0) == [0,2]

assert two_sum_1([3, 3, 4], 9) == []
```

## Решение 2: Использование словаря (Hash Map)

Данный подход применяет **словарь для хранения ранее просмотренных элементов** массива и их индексов, что позволяет находить пару чисел с заданной суммой за линейное время.

### Алгоритм
1. Создаётся пустой словарь `d`, где ключом является элемент массива, а значением — его индекс.  
2. Для каждого элемента массива `nums[i]` вычисляется разность `r = target - nums[i]`.  
3. Проверяется, присутствует ли `r` в словаре `d`.  
   - Если да, возвращаются индексы `[d[r], i]`.  
   - Если нет, текущий элемент `nums[i]` добавляется в словарь `d` с ключом `nums[i]` и значением `i`.  
4. Если ни одна пара не удовлетворяет условию, возвращается пустой список.

### Особенности
- Использование словаря позволяет мгновенно проверять наличие "комплементарного" элемента, обеспечивая **временную сложность `O(n)`**.  
- Гарантируется использование **разных индексов**, что соответствует ограничению задачи о недопустимости повторного использования одного и того же элемента.  
- Простота реализации и высокая эффективность делают этот метод предпочтительным для больших массивов.

### Вывод
Метод с использованием словаря обеспечивает линейную временную сложность, сохраняя корректность для любых размеров массива, и полностью удовлетворяет требованию задачи об оптимизации времени выполнения.

```python
def two_sum_2(nums: list[int], target: int) -> list[int]:
    d = {}
    for i in range(len(nums)):
        r = target - nums[i]
        if r in d:
            return [d[r], i]
        d[nums[i]] = i    
    return []


# Тестирование
assert two_sum_2([1, 3, 4], 4) == [0, 1]
assert two_sum_2([2, 7, 11, 15], 9) == [0,1]
assert two_sum_2([3, 2, 4], 6) == [1,2]
assert two_sum_2([3, 3], 6) == [0,1]
assert two_sum_2([-1, -2, -3, -4, -5], -8) == [2,4]
assert two_sum_2([-3, 4, 3, 90], 0) == [0,2]

assert two_sum_2([3, 3, 4], 9) == []
```

### Сравнение временной сложности алгоритмов

**Красная кривая** показывает рост времени выполнения наивного алгоритма с увеличением размера массива. Она возрастает **квадратично**, что делает метод неэффективным при больших данных.  

**Зелёная кривая** отражает алгоритм со словарём (Hash Map), где время увеличивается **линейно**, что обеспечивает высокую производительность даже при больших массивах.  

График наглядно демонстрирует преимущество алгоритма с использованием словаря по сравнению с прямым перебором, особенно при росте объёма данных.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = np.arange(1, 10, 1)  # от 1 до 10 элементов

plt.figure(figsize=(16, 8))
plt.plot(n, n**2, label='Наивный перебор O(n²)', color='red', linewidth=2)
plt.plot(n, n, label='Словарь O(n)', color='green', linewidth=2)
plt.title('Сравнение временной сложности алгоритмов Two Sum')
plt.xlabel('Размер массива n')
plt.ylabel('Условное время выполнения')
plt.legend()
plt.grid(True)
plt.show()
```
