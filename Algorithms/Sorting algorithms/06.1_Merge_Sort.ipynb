## Сортировка слиянием (Merge Sort)

Сортировка слиянием (Merge Sort) — это эффективный алгоритм сортировки, основанный на принципе **«разделяй и властвуй» (Divide and Conquer)**. Алгоритм был предложен Джоном фон Нейманом в 1945 году и до сих пор широко применяется благодаря своей стабильности и гарантированной сложности `O(n log n)`.

Основная идея заключается в том, чтобы рекурсивно разделять массив на две примерно равные части до тех пор, пока каждая часть не будет содержать один элемент. После этого соседние части последовательно объединяются в один отсортированный массив.

В отличие от большинства простых алгоритмов сортировки, Merge Sort практически не зависит от исходного порядка элементов и обеспечивает одинаково высокую производительность как на случайных, так и на почти отсортированных данных.

---

### Идея алгоритма

Пусть имеется массив

```
[8, 3, 6, 2, 7, 1, 5, 4]
```

Сначала массив рекурсивно делится пополам:

```
[8, 3, 6, 2]     [7, 1, 5, 4]
```

Затем каждая часть снова делится:

```
[8, 3] [6, 2]    [7, 1] [5, 4]
```

Разделение продолжается до массивов длины один:

```
[8] [3] [6] [2] [7] [1] [5] [4]
```

После этого начинается обратный процесс — слияние отсортированных частей:

```
[3, 8]
[2, 6]
[1, 7]
[4, 5]
```

Затем:

```
[2, 3, 6, 8]
[1, 4, 5, 7]
```

И наконец получается полностью отсортированный массив:

```
[1, 2, 3, 4, 5, 6, 7, 8]
```

---

### Преимущества

- гарантированная сложность `O(n log n)` во всех случаях;
- является стабильной сортировкой;
- хорошо подходит для сортировки связанных списков;
- эффективно работает при сортировке внешних данных (External Merge Sort);
- легко распараллеливается.

---

### Недостатки

- требует дополнительной памяти `O(n)`;
- рекурсивная реализация создаёт дополнительные вызовы функций;
- на небольших массивах обычно уступает Quick Sort и TimSort.

---

### Сложность

| Случай | Время |
|--------|-------|
| Лучший | `O(n log n)` |
| Средний | `O(n log n)` |
| Худший | `O(n log n)` |
| Память | `O(n)` |

Главным преимуществом Merge Sort является то, что время работы не зависит от исходного порядка элементов массива.

## 1) Классическая рекурсивная сортировка слиянием

Классическая реализация Merge Sort основана на рекурсивном разделении массива на две примерно равные части. Разбиение продолжается до тех пор, пока каждый подмассив не будет состоять из одного элемента, так как массив длины один уже считается отсортированным.

После завершения этапа разделения начинается процесс слияния. На каждом уровне рекурсии два соседних отсортированных подмассива объединяются в один общий отсортированный массив. Для этого элементы последовательно сравниваются, а наименьший из них помещается в результирующий массив. Когда один из подмассивов заканчивается, оставшиеся элементы второго массива просто добавляются в конец результата.

Именно эта реализация чаще всего используется для изучения алгоритма, поскольку наиболее наглядно демонстрирует принцип «разделяй и властвуй».

---

### Алгоритм

1. Если массив содержит один или ноль элементов, завершить работу.
2. Разделить массив на две примерно равные части.
3. Рекурсивно отсортировать левую часть.
4. Рекурсивно отсортировать правую часть.
5. Последовательно объединить две отсортированные части в один массив, выбирая на каждом шаге минимальный элемент.
6. Вернуть полученный отсортированный массив.

---

### Особенности

- наиболее простая и понятная реализация Merge Sort;
- полностью основана на рекурсивном принципе «разделяй и властвуй»;
- использует срезы массива, что делает код компактным и легко читаемым;
- требует дополнительной памяти как для хранения промежуточных массивов, так и для рекурсивных вызовов;
- является стабильной сортировкой.

---

### Сложность

| Параметр | Значение |
|----------|----------|
| Лучший случай | `O(n log n)` |
| Средний случай | `O(n log n)` |
| Худший случай | `O(n log n)` |
| Память | `O(n)` |

````python
def merge_sort(arr: list[int]) -> list[int]:
    if len(arr) <= 1:
        return arr

    middle = len(arr) // 2

    left = merge_sort(arr[:middle])
    right = merge_sort(arr[middle:])

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


assert merge_sort([5, 2, 4, 6, 1, 3]) == [1, 2, 3, 4, 5, 6]
assert merge_sort([9, 8, 7, 6, 5]) == [5, 6, 7, 8, 9]
assert merge_sort([1, 2, 3, 4]) == [1, 2, 3, 4]
assert merge_sort([3]) == [3]
assert merge_sort([]) == []
assert merge_sort([5, 5, 2, 2, 1, 1]) == [1, 1, 2, 2, 5, 5]
````

## 2) Рекурсивная сортировка слиянием без создания срезов

В предыдущей реализации на каждом рекурсивном вызове создавались новые подмассивы с помощью срезов (`arr[:middle]` и `arr[middle:]`). Несмотря на простоту такого подхода, создание большого количества временных списков приводит к дополнительным затратам памяти и времени.

В данной реализации массив не копируется при каждом разбиении. Вместо этого рекурсивная функция работает с границами текущего диапазона, а для слияния используется один общий вспомогательный массив, который создаётся только один раз.

Благодаря этому уменьшается количество операций копирования данных, что делает реализацию более эффективной, особенно при работе с большими массивами.

---

### Алгоритм

1. Создать один вспомогательный массив того же размера, что и исходный.
2. Рекурсивно разделять массив, передавая только границы текущего диапазона.
3. После сортировки левой и правой частей выполнить их слияние во вспомогательный массив.
4. Скопировать отсортированные элементы обратно в исходный массив.
5. После завершения рекурсии массив будет полностью отсортирован.

---

### Особенности

- не создаёт новые списки при каждом рекурсивном вызове;
- использует один общий вспомогательный массив;
- выполняет меньше операций копирования по сравнению с классической реализацией;
- сохраняет стабильность сортировки;
- чаще используется в высокопроизводительных реализациях Merge Sort.

---

### Сложность

| Параметр | Значение |
|----------|----------|
| Лучший случай | `O(n log n)` |
| Средний случай | `O(n log n)` |
| Худший случай | `O(n log n)` |
| Память | `O(n)` |

````python
def merge_sort(arr: list[int]) -> None:
    temp = [0] * len(arr)

    def sort(left: int, right: int) -> None:
        if left >= right:
            return

        middle = (left + right) // 2

        sort(left, middle)
        sort(middle + 1, right)

        i = left
        j = middle + 1
        k = left

        while i <= middle and j <= right:
            if arr[i] <= arr[j]:
                temp[k] = arr[i]
                i += 1
            else:
                temp[k] = arr[j]
                j += 1
            k += 1

        while i <= middle:
            temp[k] = arr[i]
            i += 1
            k += 1

        while j <= right:
            temp[k] = arr[j]
            j += 1
            k += 1

        for i in range(left, right + 1):
            arr[i] = temp[i]

    sort(0, len(arr) - 1)


arr = [5, 2, 4, 6, 1, 3]
merge_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
merge_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
merge_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
merge_sort(arr)
assert arr == [3]

arr = []
merge_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
merge_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## 3) Итеративная сортировка слиянием

Итеративная реализация Merge Sort, также называемая **Bottom-Up Merge Sort**, выполняет сортировку без рекурсивных вызовов.

Вместо последовательного разбиения массива на части алгоритм начинает с подмассивов длины один. Такие части уже считаются отсортированными. Затем соседние подмассивы последовательно объединяются, а их размер после каждого прохода увеличивается в два раза.

Сначала объединяются элементы по одному:

```
[5] [2] [4] [6] [1] [3]
```

Затем получаются отсортированные части длины два:

```
[2, 5] [4, 6] [1, 3]
```

После этого объединяются части длины четыре, восемь и так далее, пока весь массив не будет отсортирован.

Такой подход сохраняет основную идею Merge Sort, но полностью исключает рекурсию.

---

### Алгоритм

1. Создать вспомогательный массив того же размера, что и исходный.
2. Установить начальный размер подмассива `size = 1`.
3. Последовательно объединять соседние отсортированные части длины `size`.
4. После каждого прохода увеличивать размер частей в два раза.
5. Продолжать работу, пока `size` не станет больше или равен длине массива.

---

### Особенности

- не использует рекурсию;
- исключает затраты на рекурсивные вызовы;
- использует один вспомогательный массив;
- хорошо подходит для языков с ограниченной глубиной рекурсии;
- сохраняет стабильность сортировки;
- логика определения границ подмассивов сложнее, чем в рекурсивной реализации.

---

### Сложность

| Параметр | Значение |
|----------|----------|
| Лучший случай | `O(n log n)` |
| Средний случай | `O(n log n)` |
| Худший случай | `O(n log n)` |
| Память | `O(n)` |

````python
def merge_sort(arr: list[int]) -> None:
    n = len(arr)
    temp = [0] * n
    size = 1

    while size < n:
        left = 0

        while left < n:
            middle = min(left + size, n)
            right = min(left + 2 * size, n)

            i = left
            j = middle
            k = left

            while i < middle and j < right:
                if arr[i] <= arr[j]:
                    temp[k] = arr[i]
                    i += 1
                else:
                    temp[k] = arr[j]
                    j += 1
                k += 1

            while i < middle:
                temp[k] = arr[i]
                i += 1
                k += 1

            while j < right:
                temp[k] = arr[j]
                j += 1
                k += 1

            left += 2 * size

        arr[:] = temp[:]
        size *= 2


arr = [5, 2, 4, 6, 1, 3]
merge_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
merge_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
merge_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
merge_sort(arr)
assert arr == [3]

arr = []
merge_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
merge_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## 4) Естественная сортировка слиянием

Естественная сортировка слиянием (**Natural Merge Sort**) является развитием классического Merge Sort. Основное отличие заключается в том, что алгоритм не делит массив на части фиксированного размера, а использует уже существующие отсортированные участки массива.

Такие участки называются **сериями** (*runs*). Серия представляет собой максимальную непрерывную последовательность элементов, расположенных в неубывающем порядке. Пока каждый следующий элемент больше либо равен предыдущему, серия продолжается. Как только порядок нарушается, текущая серия заканчивается и начинается новая.

Например, для массива

```text
[1, 4, 7, 2, 5, 8, 3, 6]
```

будут найдены следующие серии:

```text
[1, 4, 7] [2, 5, 8] [3, 6]
```

Каждая из них уже отсортирована, поэтому нет необходимости разбивать её дальше. Вместо этого алгоритм сразу начинает объединять соседние серии.

После первого прохода получим:

```text
[1, 2, 4, 5, 7, 8] [3, 6]
```

После второго прохода:

```text
[1, 2, 3, 4, 5, 6, 7, 8]
```

Таким образом, алгоритм постепенно уменьшает количество серий. После каждого прохода их становится меньше, а сами серии становятся длиннее. Работа завершается, когда весь массив представляет собой одну серию, то есть полностью отсортирован.

Главное преимущество такого подхода заключается в том, что алгоритм использует уже имеющийся порядок элементов. Если массив почти отсортирован, количество серий будет небольшим, поэтому потребуется значительно меньше операций слияния, чем в классическом Merge Sort.

---

### Алгоритм

1. Просмотреть массив слева направо и найти все максимальные неубывающие серии.
2. Запомнить границы каждой найденной серии.
3. Последовательно объединить первую и вторую серии, затем третью и четвёртую и так далее.
4. Если количество серий оказалось нечётным, последняя серия без изменений переносится в следующий проход.
5. После завершения слияния выполнить новый поиск серий.
6. Повторять процесс до тех пор, пока весь массив не станет одной серией.

---

### Особенности

- использует уже существующие отсортированные участки массива;
- не выполняет искусственное разбиение массива пополам;
- на почти отсортированных данных может выполнить значительно меньше операций слияния;
- не использует рекурсивное разбиение массива;
- сохраняет стабильность сортировки;
- сложнее в реализации, поскольку необходимо определять границы всех серий перед каждым проходом;
- в худшем случае имеет ту же асимптотическую сложность, что и классический Merge Sort.

---

### Сложность

| Параметр | Значение |
|----------|----------|
| Лучший случай | `O(n)` |
| Средний случай | `O(n log n)` |
| Худший случай | `O(n log n)` |
| Память | `O(n)` |

Лучший случай достигается, когда массив уже отсортирован. В этом случае весь массив является одной серией, поэтому алгоритму достаточно одного линейного прохода для проверки порядка элементов.

````python
def merge_sort(arr: list[int]) -> None:
    n = len(arr)

    if n <= 1:
        return

    temp = [0] * n

    while True:
        runs = []
        left = 0

        while left < n:
            right = left + 1

            while right < n and arr[right - 1] <= arr[right]:
                right += 1

            runs.append((left, right))
            left = right

        if len(runs) == 1:
            return

        for index in range(0, len(runs), 2):
            left, middle = runs[index]

            if index + 1 == len(runs):
                temp[left:middle] = arr[left:middle]
                continue

            _, right = runs[index + 1]

            i = left
            j = middle
            k = left

            while i < middle and j < right:
                if arr[i] <= arr[j]:
                    temp[k] = arr[i]
                    i += 1
                else:
                    temp[k] = arr[j]
                    j += 1
                k += 1

            while i < middle:
                temp[k] = arr[i]
                i += 1
                k += 1

            while j < right:
                temp[k] = arr[j]
                j += 1
                k += 1

        arr[:] = temp[:]

arr = [5, 2, 4, 6, 1, 3]
merge_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
merge_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
merge_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
merge_sort(arr)
assert arr == [3]

arr = []
merge_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
merge_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]

arr = [1, 4, 7, 2, 5, 8, 3, 6]
merge_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6, 7, 8]
````

## Вывод

Были рассмотрены основные реализации сортировки слиянием, начиная с классического рекурсивного алгоритма и заканчивая его более эффективными вариантами.

Классическая реализация позволяет лучше всего понять принцип «разделяй и властвуй», лежащий в основе Merge Sort. Далее была рассмотрена реализация без создания срезов, которая уменьшает количество выделений памяти и лучше подходит для практического применения. Затем изучена итеративная версия (**Bottom-Up Merge Sort**), демонстрирующая, что алгоритм может быть реализован без использования рекурсии. В завершение была рассмотрена естественная сортировка слиянием (**Natural Merge Sort**), использующая уже существующие отсортированные серии и способная работать значительно эффективнее на почти отсортированных данных.

Несмотря на различия в реализации, все рассмотренные алгоритмы сохраняют ключевые преимущества Merge Sort: гарантированную сложность `O(n log n)` в худшем случае, стабильность сортировки и предсказуемое время работы независимо от входных данных. Основным недостатком остаётся необходимость использования дополнительной памяти объёмом `O(n)`.

Помимо рассмотренных реализаций существуют и другие разновидности Merge Sort, такие как **In-Place Merge Sort**, **External Merge Sort**, **Parallel Merge Sort** и **K-Way Merge**. Каждая из них предназначена для решения специализированных задач: уменьшения потребления памяти, обработки больших объёмов данных, выполнения параллельных вычислений или объединения большого количества отсортированных последовательностей.
