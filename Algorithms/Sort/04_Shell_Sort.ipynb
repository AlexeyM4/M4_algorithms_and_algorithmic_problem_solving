## Сортировка Шелла (Shell Sort)

Сортировка Шелла (Shell Sort) — это улучшенная версия сортировки вставками, предложенная Дональдом Шеллом в 1959 году.

Главная идея заключается в том, чтобы сначала сортировать элементы, находящиеся далеко друг от друга, постепенно уменьшая расстояние между ними. Благодаря этому к моменту последнего прохода массив уже оказывается почти отсортированным, поэтому финальная сортировка вставками выполняется значительно быстрее.

В отличие от обычной сортировки вставками, которая перемещает элементы только на одну позицию за раз, сортировка Шелла позволяет элементам быстро перемещаться на большие расстояния.

---

### Идея алгоритма

Пусть имеется массив

```
[8, 5, 3, 7, 6, 2, 1, 4]
```

Если выбрать шаг (`gap`) равный 4, то массив разбивается на несколько независимых подпоследовательностей:

```
8 6
5 2
3 1
7 4
```

Каждая из них сортируется вставками.

После этого шаг уменьшается:

```
gap = 2
```

Получаются новые группы.

Когда шаг становится равным единице, выполняется обычная сортировка вставками.

Так как массив уже почти отсортирован, последний проход выполняется очень быстро.

---

### Преимущества

- значительно быстрее обычной сортировки вставками;
- не требует дополнительной памяти;
- проста в реализации;
- хорошо работает на массивах среднего размера.

---

### Недостатки

- не является стабильной;
- эффективность сильно зависит от выбора последовательности шагов;
- асимптотика хуже современных алгоритмов (Quick Sort, Merge Sort, TimSort).

---

### Сложность

| Случай | Время |
|--------|-------|
| Лучший | O(n log n) *(для удачных последовательностей шагов)* |
| Средний | зависит от последовательности шагов |
| Худший | O(n²) |
| Память | O(1) |

Точная сложность алгоритма зависит от выбранной последовательности шагов.

## 1) Классическая сортировка Шелла

Классическая реализация сортировки Шелла использует простую последовательность шагов: на каждой итерации расстояние между сравниваемыми элементами уменьшается в два раза.

Несмотря на простоту, именно эта реализация положила начало целому семейству алгоритмов сортировки Шелла.

---

### Алгоритм

1. Вычислить начальный шаг `gap = n // 2`.
2. Пока `gap > 0`:
   - пройти по массиву, начиная с индекса `gap`;
   - для каждого элемента выполнить сортировку вставками с текущим шагом `gap`;
   - после завершения прохода уменьшить шаг вдвое.
3. После прохода с шагом `1` массив будет полностью отсортирован.

---

### Особенности

- не требует дополнительной памяти;
- основана на сортировке вставками;
- значительно быстрее обычной сортировки вставками на большинстве массивов;
- эффективность зависит от выбора последовательности шагов;
- не является стабильной.

---

### Сложность

| Параметр | Значение |
|----------|----------|
| Лучший случай | зависит от последовательности шагов |
| Средний случай | примерно `O(n^1.5)` |
| Худший случай | `O(n²)` |
| Память | `O(1)` |

````python
def shell_sort(arr: list[int]) -> None:
    gap = len(arr) // 2

    while gap > 0:
        for i in range(gap, len(arr)):
            current = arr[i]
            j = i

            while j >= gap and arr[j - gap] > current:
                arr[j] = arr[j - gap]
                j -= gap

            arr[j] = current

        gap //= 2


arr = [5, 2, 4, 6, 1, 3]
shell_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
shell_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
shell_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
shell_sort(arr)
assert arr == [3]

arr = []
shell_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
shell_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## 2) Сортировка Шелла с последовательностью Кнута

В этом варианте используется последовательность шагов, предложенная Дональдом Кнутом:

```
1, 4, 13, 40, 121, ...
```

Каждый следующий шаг вычисляется по формуле:

```
gap = gap * 3 + 1
```

По сравнению с классической последовательностью такой подход обеспечивает более равномерное уменьшение шага и на практике часто работает быстрее.

---

### Алгоритм

1. Построить максимальный шаг по последовательности Кнута, не превышающий размер массива.
2. Выполнять сортировку вставками с текущим шагом `gap`.
3. После каждого прохода уменьшать шаг по формуле `gap //= 3`.
4. Завершить работу, когда шаг станет равен `1`.

---

### Особенности

- одна из самых популярных последовательностей шагов;
- обычно работает быстрее классического варианта;
- проста в реализации.

````python
def shell_sort(arr: list[int]) -> None:
    n = len(arr)

    gap = 1
    while gap < n // 3:
        gap = gap * 3 + 1

    while gap > 0:
        for i in range(gap, n):
            current = arr[i]
            j = i

            while j >= gap and arr[j - gap] > current:
                arr[j] = arr[j - gap]
                j -= gap

            arr[j] = current

        gap //= 3

arr = [5, 2, 4, 6, 1, 3]
shell_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
shell_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
shell_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
shell_sort(arr)
assert arr == [3]

arr = []
shell_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
shell_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## 3) Сортировка Шелла с последовательностью Хиббарда

В этом варианте используется последовательность шагов, предложенная Томасом Хиббардом:

```
1, 3, 7, 15, 31, ...
```

Каждый шаг вычисляется по формуле:

```
gap = 2^k - 1
```

По сравнению с классической последовательностью и последовательностью Кнута данный вариант обеспечивает более удачное распределение шагов, что позволяет уменьшить количество сравнений и перемещений элементов.

---

### Алгоритм

1. Построить последовательность Хиббарда, не превышающую размер массива.
2. Выполнять сортировку вставками с текущим шагом `gap`, начиная с наибольшего.
3. Последовательно переходить к следующему меньшему шагу.
4. После прохода с шагом `1` массив будет полностью отсортирован.

---

### Особенности

- использует последовательность вида `2^k - 1`;
- выполняет меньше сравнений по сравнению с классической последовательностью;
- обеспечивает хорошую производительность на практике;
- проста в реализации.

````python
def shell_sort(arr: list[int]) -> None:
    n = len(arr)

    gaps = []
    gap = 1
    while gap < n:
        gaps.append(gap)
        gap = gap * 2 + 1

    for gap in reversed(gaps):
        for i in range(gap, n):
            current = arr[i]
            j = i

            while j >= gap and arr[j - gap] > current:
                arr[j] = arr[j - gap]
                j -= gap

            arr[j] = current

arr = [5, 2, 4, 6, 1, 3]
shell_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
shell_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
shell_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
shell_sort(arr)
assert arr == [3]

arr = []
shell_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
shell_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## 4) Сортировка Шелла с последовательностью Седжвика

В этом варианте используется последовательность шагов, предложенная Робертом Седжвиком. Она была разработана для уменьшения количества сравнений и перемещений элементов по сравнению с более ранними последовательностями.

Шаги вычисляются по формулам:

Для чётных `k`:

```
gap = 9 · (2^k − 2^(k / 2)) + 1
```

Для нечётных `k`:

```
gap = 8 · 2^k − 6 · 2^((k + 1) / 2) + 1
```

Первые значения последовательности:

```
1, 5, 19, 41, 109, 209, 505, 929, ...
```

На практике данный вариант часто показывает одну из лучших производительностей среди классических реализаций сортировки Шелла.

---

### Алгоритм

1. Построить последовательность Седжвика, не превышающую размер массива.
2. Выполнять сортировку вставками с текущим шагом `gap`, начиная с наибольшего.
3. Последовательно переходить к следующему меньшему шагу.
4. После прохода с шагом `1` массив будет полностью отсортирован.

---

### Особенности

- использует две формулы для построения последовательности шагов;
- одна из наиболее эффективных классических последовательностей;
- уменьшает количество сравнений и перемещений элементов;
- хорошо подходит для массивов среднего и большого размера.

````python
def shell_sort(arr: list[int]) -> None:
    n = len(arr)

    gaps = []
    k = 0

    while True:
        if k % 2 == 0:
            gap = 9 * (2**k - 2 ** (k // 2)) + 1
        else:
            gap = 8 * 2**k - 6 * 2 ** ((k + 1) // 2) + 1

        if gap >= n:
            break

        gaps.append(gap)
        k += 1

    for gap in reversed(gaps):
        for i in range(gap, n):
            current = arr[i]
            j = i

            while j >= gap and arr[j - gap] > current:
                arr[j] = arr[j - gap]
                j -= gap

            arr[j] = current


arr = [5, 2, 4, 6, 1, 3]
shell_sort(arr)
assert arr == [1, 2, 3, 4, 5, 6]

arr = [9, 8, 7, 6, 5]
shell_sort(arr)
assert arr == [5, 6, 7, 8, 9]

arr = [1, 2, 3, 4]
shell_sort(arr)
assert arr == [1, 2, 3, 4]

arr = [3]
shell_sort(arr)
assert arr == [3]

arr = []
shell_sort(arr)
assert arr == []

arr = [5, 5, 2, 2, 1, 1]
shell_sort(arr)
assert arr == [1, 1, 2, 2, 5, 5]
````

## Другие последовательности шагов

Помимо рассмотренных вариантов существует множество других последовательностей шагов для сортировки Шелла. Основная идея всех этих последовательностей одинакова — подобрать значения `gap` таким образом, чтобы уменьшить количество сравнений и перемещений элементов.

Наиболее известные из них:

### Последовательность Пратта

Использует все числа вида

```
gap = 2^i · 3^j
```

Например:

```
1, 2, 3, 4, 6, 8, 9, 12, 16, 18, 24, ...
```

---

### Последовательность Папернова — Стасевича

Вычисляется по формуле

```
gap = 2^k + 1
```

Первые значения:

```
3, 5, 9, 17, 33, 65, ...
```

---

### Последовательность Гоннета (Gonnet)

Следующий шаг вычисляется как

```
gap = ⌊gap / 2.2⌋
```

где начальный шаг обычно равен размеру массива.

---

### Последовательность Токуды (Tokuda)

Вычисляется по формуле

```
gap = ⌈(9 · (9 / 4)^k - 4) / 5⌉
```

Первые значения:

```
1, 4, 9, 20, 46, 103, 233, 525, ...
```

---

### Последовательность Циуры (Ciura)

Основана на экспериментально подобранных значениях:

```
1, 4, 10, 23, 57, 132, 301, 701
```

Для больших массивов последовательность обычно продолжают по правилу

```
gap = ⌊gap · 2.25⌋
```

Несмотря на отсутствие строгого математического обоснования, именно эта последовательность часто демонстрирует лучшие результаты на практике.

---

В настоящее время не существует единственной последовательности шагов, которая была бы оптимальной для всех случаев. Поэтому выбор последовательности зависит от конкретной задачи, размера массива и требований к производительности.
