### Задача 1

Рассмотрим три числа: a, b и c. Упорядочим их по возрастанию

<b>Вопрос:</b>
какое число будет стоять между двумя другими?

<b>Формат ввода:</b><br>
В единственной строке записаны три целых числа a, b и c (−1000 ≤ a, b, c ≤ 1000), числа разделены пробелами

<b>Формат вывода:</b><br>
Выведите число, которое будет стоять между двумя другими после упорядочивания


### Решение

Считываются три целых числа, введённые через пробел:<br>

Функция input() получает строку<br> 
Метод split() разбивает её на отдельные элементы через знак пробела<br>
Функция map(int, …) преобразует каждую часть в целое число<br>

Результат присваивается переменным a, b и c

```python
a, b, c = map(int, input().split())
```

**Алгоритм 1 "Сортировка"**

Создаётся список [a, b, c]<br>
Функция sorted() возвращает новый список, отсортированный по возрастанию<br>
Обращение к элементу с индексом [1] даёт второй элемент отсортированного списка, то есть среднее значение (число, стоящее между двумя другими)<br>

```python
print(sorted([a, b, c])[1])
```

**Алгоритм 2 "Медиана"**

Импортируется функция median из модуля statistics, которая вычисляет медиану (среднее значение) числового набора

Создаётся список [a, b, c]<br>
Функция median() вычисляет среднее значение списка, то есть число, которое находится между двумя другими

```python
from statistics import median

print(median([a, b, c]))
```

**Алгоритм 3 "Сумма минус мин/макс"**

Создаётся список nums, содержащий три числа: a, b и c<br>

sum(nums) вычисляет сумму всех чисел списка<br>
max(nums) находит наибольшее число, min(nums) — наименьшее<br>
Вычитание максимального и минимального значения из суммы даёт среднее число, стоящее между двумя другими

```python
nums = [a, b, c]

print(sum(nums) - max(nums) - min(nums))
```

**Алгоритм 4 "Условия"**

Последовательно проверяет, является ли одно из чисел средним, используя операторы сравнения.

1.  Проверка для **b**: `(a <= b <= c) or (c <= b <= a)`
2.  Проверка для **a**: `(b <= a <= c) or (c <= a <= b)`
3.  Если ни одно из условий не сработало, среднее — **c**.

```python
if (a <= b <= c) or (c <= b <= a):
    print(b)
elif (b <= a <= c) or (c <= a <= b):
    print(a)
else:
    print(c)
```
