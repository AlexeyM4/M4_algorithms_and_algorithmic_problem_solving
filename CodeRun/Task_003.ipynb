## Задача 3. Вывести маршрут максимальной стоимости

В левом верхнем углу прямоугольной таблицы размером **`N × M`** находится черепашка.

Каждая клетка таблицы содержит некоторое целое число. За один ход черепашка может переместиться только:

- вправо (`R`);
- вниз (`D`).

Перемещаться **влево** и **вверх** запрещено.

Маршрут начинается в **левой верхней** клетке и заканчивается в **правой нижней**.

Стоимость маршрута равна сумме чисел во всех клетках, через которые проходит черепашка, включая начальную и конечную клетки.

Необходимо определить:

1. максимальную возможную стоимость маршрута;
2. один из маршрутов, на котором эта стоимость достигается.

---

## Формат ввода

В первой строке записаны два натуральных числа:

```text
N M
```

где:

- `1 ≤ N ≤ 100` — количество строк;
- `1 ≤ M ≤ 100` — количество столбцов.

Далее следуют `N` строк, каждая из которых содержит `M` целых чисел — значения, записанные в клетках таблицы.

---

## Формат вывода

В первой строке выведите максимальную возможную стоимость маршрута.

Во второй строке выведите маршрут, состоящий из символов:

- `D` — переход вниз;
- `R` — переход вправо.

Маршрут должен содержать:

- `N - 1` символов `D`;
- `M - 1` символов `R`.

Если существует несколько оптимальных маршрутов, разрешается вывести любой из них.

---

## Ограничения

- `1 ≤ N, M ≤ 100`
- `0 ≤ a[i][j] ≤ 100`
- Ограничение времени — **1 секунда**.
- Ограничение памяти — **256 МБ**.

---

## Пример

**Ввод**

```text
5 5
9 9 9 9 9
3 0 0 0 0
9 9 9 9 9
6 6 6 6 8
9 9 9 9 9
```

**Вывод**

```text
74
D D R R R R D D
```

## 1) Динамическое программирование с восстановлением маршрута

Для каждой клетки таблицы вычисляется максимальная стоимость пути из левого верхнего угла до этой клетки.

Так как черепашка может двигаться только вправо и вниз, попасть в клетку `(i, j)` можно двумя способами:

- сверху — из клетки `(i - 1, j)`;
- слева — из клетки `(i, j - 1)`.

В текущую клетку прибавляется большее из двух накопленных значений. Одновременно в словаре `d` сохраняется направление перехода:

- `D` — черепашка пришла сверху, сделав шаг вниз;
- `R` — черепашка пришла слева, сделав шаг вправо.

После заполнения таблицы маршрут восстанавливается от правой нижней клетки к левой верхней.

### Алгоритм

1. Обработать первую строку:
   - накопить значения слева направо;
   - для каждой клетки сохранить направление `R`.
2. Обработать первый столбец:
   - накопить значения сверху вниз;
   - для каждой клетки сохранить направление `D`.
3. Для каждой оставшейся клетки:
   - сравнить накопленную стоимость сверху и слева;
   - выбрать большее значение;
   - прибавить его к текущей клетке;
   - сохранить направление выбранного перехода.
4. Начать восстановление маршрута из клетки `(N - 1, M - 1)`.
5. Пока не достигнута клетка `(0, 0)`:
   - получить сохранённое направление;
   - добавить его в маршрут;
   - при `R` перейти на один столбец влево;
   - при `D` перейти на одну строку вверх.
6. Развернуть полученный маршрут, так как он был построен от конца к началу.

### Особенности

- Исходная матрица используется для хранения максимальных накопленных сумм.
- Словарь `d` хранит направление перехода для каждой клетки.
- Если значения сверху и слева равны, выбирается переход сверху.
- При нескольких оптимальных маршрутах выводится один из них.
- Маршрут восстанавливается циклом `while`, который завершается после достижения начальной клетки.
- Временная сложность — **O(N · M)**.
- Дополнительная память — **O(N · M)**.

````python
n, m = 5, 5

matrix = [
    [9, 9, 9, 9, 9],
    [3, 0, 0, 0, 0],
    [9, 9, 9, 9, 9],
    [6, 6, 6, 6, 8],
    [9, 9, 9, 9, 9]
]

d = {}

for i in range(1, m):
    matrix[0][i] +=  matrix[0][i - 1]
    d[(0, i)] = 'R'
    
    
for i in range(1, n):
    matrix[i][0] +=  matrix[i - 1][0]
    d[(i, 0)] = 'D'

    for j in range(1, m):
        if matrix[i - 1][j] >= matrix[i][j - 1]:
            matrix[i][j] += matrix[i - 1][j]
            d[(i, j)] = 'D'
        else:
            matrix[i][j] += matrix[i][j - 1]
            d[(i, j)] = 'R'

result = []

i, j = n - 1, m - 1
while i > 0 or j > 0:
    direction = d[(i, j)]
    result.append(direction)

    if direction == 'R':
        j -= 1
    else:
        i -= 1
        
print(matrix[n - 1][m - 1])
print(*result[::-1])
````

## 2) Динамическое программирование с матрицей направлений

Как и в предыдущем решении, для каждой клетки вычисляется максимальная стоимость пути от левого верхнего угла.

Отличие заключается в способе хранения информации для восстановления маршрута. Вместо словаря используется отдельная матрица `parent`, в которой для каждой клетки сохраняется направление перехода:

- `D` — переход выполнен сверху;
- `R` — переход выполнен слева.

После заполнения таблицы маршрут восстанавливается, начиная с правой нижней клетки и последовательно переходя к предыдущим клеткам согласно значениям матрицы `parent`.

### Алгоритм

1. Создать матрицу `parent` размером `N × M`.
2. Обработать первую строку:
   - накопить максимальные стоимости;
   - сохранить направление `R`.
3. Обработать первый столбец:
   - накопить максимальные стоимости;
   - сохранить направление `D`.
4. Для каждой оставшейся клетки:
   - выбрать большее значение между верхней и левой клетками;
   - прибавить его к стоимости текущей клетки;
   - записать соответствующее направление в матрицу `parent`.
5. Начать восстановление маршрута из клетки `(N - 1, M - 1)`.
6. Пока не достигнута клетка `(0, 0)`:
   - считать направление из `parent`;
   - добавить его в маршрут;
   - перейти в предыдущую клетку.
7. Развернуть полученный маршрут и вывести ответ.

### Особенности

- Для восстановления маршрута используется матрица направлений вместо словаря.
- Матрица направлений имеет более простую структуру данных и требует меньше памяти, чем словарь с ключами-кортежами.
- Если значения сверху и слева равны, выбирается переход сверху.
- При нескольких оптимальных маршрутах выводится один из них.
- Временная сложность — **O(N · M)**.
- Дополнительная память — **O(N · M)**.

````python
n, m = 5, 5

matrix = [
    [9, 9, 9, 9, 9],
    [3, 0, 0, 0, 0],
    [9, 9, 9, 9, 9],
    [6, 6, 6, 6, 8],
    [9, 9, 9, 9, 9]
]

parent = [[''] * m for _ in range(n)]

for j in range(1, m):
    matrix[0][j] += matrix[0][j - 1]
    parent[0][j] = 'R'

for i in range(1, n):
    matrix[i][0] += matrix[i - 1][0]
    parent[i][0] = 'D'

    for j in range(1, m):
        if matrix[i - 1][j] >= matrix[i][j - 1]:
            matrix[i][j] += matrix[i - 1][j]
            parent[i][j] = 'D'
        else:
            matrix[i][j] += matrix[i][j - 1]
            parent[i][j] = 'R'

route = []

i, j = n - 1, m - 1
while i > 0 or j > 0:
    direction = parent[i][j]
    route.append(direction)

    if direction == 'D':
        i -= 1
    else:
        j -= 1

print(matrix[-1][-1])
print(*reversed(route))
````

## Вывод

В данной задаче были рассмотрены два способа восстановления маршрута после вычисления максимальной стоимости пути с помощью динамического программирования.

В первом решении информация о предыдущем переходе хранится в словаре, что делает восстановление маршрута простым и наглядным.

Во втором решении вместо словаря используется отдельная матрица направлений. Такой подход также имеет сложность **O(N · M)**, но использует более компактную структуру данных и лучше подходит для задач, где информация хранится для каждой клетки таблицы.

Оба решения имеют одинаковую асимптотическую сложность:

- время — **O(N · M)**;
- дополнительная память — **O(N · M)**.

На практике чаще используют второй вариант, поскольку матрица направлений лучше соответствует структуре задачи и делает код более простым и понятным.
