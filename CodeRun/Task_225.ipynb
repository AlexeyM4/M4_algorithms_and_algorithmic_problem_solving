## Задача 225. Дипломы

У Пети есть `n` одинаковых дипломов размером `w × h`.

Он хочет разместить все дипломы на квадратной доске минимального размера. Дипломы нельзя поворачивать на `90°`, а области, занимаемые разными дипломами, не должны пересекаться.

Необходимо определить минимальную длину стороны квадратной доски, на которой можно разместить все `n` дипломов.

---

### Формат ввода

В единственной строке записаны три целых числа:

- `w` — ширина одного диплома;
- `h` — высота одного диплома;
- `n` — количество дипломов.

---

### Формат вывода

Выведите одно целое число — минимальную длину стороны квадратной доски, на которой можно разместить все `n` дипломов.

---

### Ограничения

- `1 ≤ w, h, n ≤ 10⁹`;
- дипломы нельзя поворачивать;
- ограничение времени — 1 секунда;
- ограничение памяти — 64 МБ.

---

### Пример

**Ввод**

```
2 3 10
```

**Вывод**

```
9
```

На квадратной доске со стороной `9` можно разместить:

- `9 // 2 = 4` диплома по горизонтали;
- `9 // 3 = 3` диплома по вертикали.

Всего помещается:

```python
4 * 3 = 12
```

Этого достаточно для размещения `10` дипломов. Квадрат меньшего размера не позволяет разместить их все.

## 1) Решение с помощью бинарного поиска по ответу

Размер стороны квадрата заранее неизвестен, однако можно заметить, что если на квадрате со стороной `x` удаётся разместить все дипломы, то на любом квадрате большего размера это также будет возможно.

Это позволяет применить **бинарный поиск по ответу** и найти минимальную подходящую длину стороны квадрата.

---

### Алгоритм

1. Задать диапазон поиска возможного ответа.
2. Найти середину диапазона.
3. Проверить, можно ли разместить все дипломы на квадрате с такой стороной.
4. Если можно — попробовать квадрат меньшего размера.
5. Иначе увеличить размер квадрата.
6. Повторять, пока не будет найден минимальный возможный ответ.

---

### Проверка (`check`)

Функция получает длину стороны квадрата `x` и определяет, можно ли разместить на нём все дипломы.

Сначала вычисляется количество строк, которое помещается по высоте квадрата.

```
rows = x // h
```

Затем определяется, сколько столбцов потребуется для размещения всех дипломов.

```
columns = ceil(n / rows)
```

После этого вычисляется необходимая ширина.

```
columns * w <= x
```

Если ширина не превышает сторону квадрата, значит все дипломы помещаются.

---

### Бинарный поиск

Левая граница поиска равна большему из размеров диплома.

```
left = max(w, h)
```

Правая граница — размещение всех дипломов в один ряд.

```
right = left * n
```

Во время бинарного поиска проверяется середина диапазона.

Если разместить дипломы удалось, правая граница сдвигается влево.

```
right = mid
```

Иначе ответ находится правее.

```
left = mid + 1
```

После завершения поиска обе границы совпадут и будут содержать минимальную длину стороны квадрата.

---

### Особенности

- используется бинарный поиск по ответу;
- каждая проверка выполняется за `O(1)`;
- решение не перебирает возможные размеры квадрата.

---

### Сложность

Каждая проверка выполняется за `O(1)`, а бинарный поиск делает `O(log(n · max(w, h)))` проверок.

- **Время:** `O(log(n · max(w, h)))`.
- **Дополнительная память:** `O(1)`.

````python
from math import ceil

def check(w: int, h: int, n: int, x: int) -> bool:
    rows = x // h
    columns = ceil(n / rows)
    return columns * w <= x

def main():
    # w, h, n = map(int, input().split())
    w, h, n = 2, 3, 10

    left = max(w, h)
    right = left * n
    while left < right:
        mid = (left + right) // 2
        if check(w, h, n, mid):
            right = mid
        else:
            left = mid + 1
    return right


main()
````

## 2) Решение с помощью модуля `bisect`

Вместо собственной реализации бинарного поиска можно воспользоваться функцией `bisect_left` из стандартного модуля `bisect`.

Функция позволяет найти первое значение стороны квадрата, при котором проверка `check` возвращает `True`, то есть все дипломы помещаются на доске.

---

### Алгоритм

1. Задать диапазон возможных ответов.
2. Для каждого размера квадрата выполнять проверку с помощью функции `check`.
3. Найти первое значение, для которого проверка возвращает `True`.
4. Вывести найденную минимальную длину стороны квадрата.

---

### Поиск ответа

Для поиска используется функция `bisect_left`.

```
answer = bisect_left(range(left, right + 1),
                     True,
                     key=lambda i: check(w, h, n, i))
```

Для каждого значения из диапазона вызывается функция `check`. Как только результат проверки становится `True`, `bisect_left` находит индекс первого такого элемента.

Поскольку функция возвращает индекс внутри диапазона `range(left, right + 1)`, итоговый ответ вычисляется следующим образом:

```
left + answer
```

---

### Особенности

- используется готовая реализация бинарного поиска;
- функция `check` остаётся без изменений;
- код становится немного компактнее, сохраняя ту же эффективность.

---

### Сложность

Каждая проверка выполняется за `O(1)`, а бинарный поиск делает `O(log(n · max(w, h)))` проверок.

- **Время:** `O(log(n · max(w, h)))`.
- **Дополнительная память:** `O(1)`.

````python
from bisect import bisect_left
from math import ceil


def check(w: int, h: int, n: int, x: int) -> bool:
    rows = x // h
    columns = ceil(n / rows)
    return columns * w <= x


def main():
    # w, h, n = map(int, input().split())
    w, h, n = 2, 3, 10

    left = max(w, h)
    right = left * n

    answer = bisect_left(range(left, right + 1),
                         True,
                         key=lambda i: check(w, h, n, i))

    return left + answer


print(main())
````

## Вывод

В данной задаче требовалось найти минимальный размер квадратной доски, на которой можно разместить все дипломы.

Для решения был использован **бинарный поиск по ответу**. Вместо перебора всех возможных размеров квадрата проверяется только логарифмическое количество вариантов, что позволяет эффективно работать даже при очень больших ограничениях.

Были рассмотрены два варианта реализации. Первый выполняет бинарный поиск вручную, что помогает лучше понять принцип работы алгоритма. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, позволяя сделать код немного компактнее. Оба решения имеют одинаковую асимптотическую сложность и эффективно решают поставленную задачу.
