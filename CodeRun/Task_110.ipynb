## 1) Решение с помощью бинарного поиска

Так как требуется найти **ближайшее** каноническое значение для каждого запроса, сначала отсортируем массив канонических значений. После сортировки ближайший элемент можно находить с помощью бинарного поиска за `O(log n)`.

Идея заключается в том, чтобы найти два соседних элемента, между которыми находится искомое значение. Именно один из них и будет ближайшим.

---

### Алгоритм

1. Считать все канонические значения.
2. Отсортировать их по возрастанию.
3. Для каждого запроса выполнить бинарный поиск.
4. Найти два соседних элемента, между которыми находится искомое число.
5. Сравнить расстояния до этих элементов.
6. Вывести ближайшее каноническое значение.

---

### Бинарный поиск (`bin_search`)

Сначала рассматриваются два простых случая.
- Если искомое значение меньше либо равно минимальному элементу массива, ближайшим сразу является первый элемент;
- Если искомое значение больше либо равно максимальному элементу, ближайшим будет последний элемент массива;
- Во всех остальных случаях выполняется бинарный поиск.

Во время поиска поддерживаются два указателя:
- `left` — элемент, который меньше искомого значения;
- `right` — элемент, который больше либо равен искомому значению.

Поиск продолжается до тех пор, пока между указателями не останется соседняя пара элементов.

После завершения бинарного поиска остаются только два возможных кандидата — `arr[left]` и `arr[right]`. Остаётся сравнить расстояния до искомого значения и вернуть ближайший элемент.

```
if abs(t - arr[left]) < abs(t - arr[right]):
    return arr[left]
return arr[right]
```

Если расстояния одинаковы, задача разрешает вывести любой из вариантов, поэтому можно вернуть любой из двух элементов.

---

### Особенности

- массив канонических значений сортируется только один раз;
- каждый запрос обрабатывается независимо;
- для каждого запроса используется бинарный поиск;
- после поиска достаточно сравнить только два соседних элемента.

---

### Сложность

- Сортировка массива: `O(n log n)`.
- Обработка одного запроса: `O(log n)`.

Итоговая сложность алгоритма:

- **Время:** `O(n log n + m log n)`.
- **Дополнительная память:** `O(1)` (без учёта сортировки).

````python
def bin_search(arr: list[int], t: int) -> int:
    left = 0
    right = len(arr) - 1

    if t <= arr[left]:
        return arr[left]
    elif arr[right] <= t:
        return arr[right]

    while right - left > 1:
        m = (left + right) // 2
        if arr[m] < t:
            left = m
        else:
            right = m

    if abs(t - arr[left]) < abs(t - arr[right]):
        return arr[left]
    return arr[right]


def main():
    # n = int(input())
    # lst_n = list(map(int, input().split()))
    # m = int(input())
    # lst_m = [int(input()) for _ in range(m)]

    n = 6
    lst_n = [-7, 3, 8, 12, -3, -10]
    m = 7
    lst_m = [-12, 13, -5, -3, 5, -2, 0]

    lst_n.sort()

    for i in lst_m:
        print(bin_search(lst_n, i))


main()
````

## 2) Решение с помощью модуля `bisect`

В Python уже существует готовая реализация бинарного поиска — модуль `bisect`. Он позволяет быстро находить позицию элемента в отсортированном массиве, избавляя от необходимости самостоятельно реализовывать бинарный поиск.

В данном решении используется функция `bisect_left`, которая возвращает индекс, куда следует вставить число, чтобы сохранить порядок сортировки.

---

### Алгоритм

1. Отсортировать массив канонических значений.
2. Для каждого запроса вызвать `bisect_left`.
3. Получить позицию, куда должен быть вставлен искомый элемент.
4. Рассмотреть два ближайших кандидата:
   - элемент слева;
   - элемент справа.
5. Сравнить расстояния до этих элементов.
6. Вывести ближайшее каноническое значение.

---

### Поиск ближайшего элемента (`nearest`)

Функция `bisect_left` возвращает индекс первого элемента, который **не меньше** искомого значения.

```
pos = bisect_left(arr, t)
```

После этого возможны три случая.
- Если `pos == 0`, искомое значение находится левее всех элементов массива, поэтому ближайшим является первый элемент.
- Если `pos == len(arr)`, искомое значение больше всех элементов массива, поэтому ближайшим является последний элемент.

Во всех остальных случаях искомое значение располагается между двумя соседними элементами массива:
- `arr[pos - 1]`;
- `arr[pos]`.

Остаётся сравнить расстояния до этих элементов и вернуть ближайший.

```
if abs(t - arr[pos - 1]) < abs(t - arr[pos]):
    return arr[pos - 1]
return arr[pos]
```

Если расстояния одинаковы, задача разрешает вывести любой из двух вариантов.

---

### Особенности

- используется готовая реализация бинарного поиска из стандартной библиотеки;
- код получается компактнее и проще;
- для каждого запроса рассматриваются только два ближайших элемента;
- асимптотика полностью совпадает с ручной реализацией бинарного поиска.

---

### Сложность

- Сортировка массива: `O(n log n)`.
- Обработка одного запроса: `O(log n)`.

Итоговая сложность алгоритма:

- **Время:** `O(n log n + m log n)`.
- **Дополнительная память:** `O(1)` (без учёта сортировки).

````python
from bisect import bisect_left


def nearest(arr: list[int], t: int) -> int:
    pos = bisect_left(arr, t)

    if pos == 0:
        return arr[0]

    if pos == len(arr):
        return arr[-1]

    if abs(t - arr[pos - 1]) < abs(t - arr[pos]):
        return arr[pos - 1]
    return arr[pos]

def main():
    # n = int(input())
    # list_n = sorted(map(int, input().split()))
    # m = int(input())
    # lst_m = int(input())

    n = 6
    lst_n = [-7, 3, 8, 12, -3, -10]
    m = 7
    lst_m = [-12, 13, -5, -3, 5, -2, 0]

    lst_n.sort()

    for i in lst_m:
        print(nearest(lst_n, i))

main()
````

## Вывод

В данной задаче требовалось для каждого значения найти ближайший элемент в заданном наборе чисел. Благодаря предварительной сортировке массива эту задачу удалось эффективно решить с помощью **бинарного поиска**, что значительно быстрее, чем последовательный просмотр всех элементов для каждого запроса.

Были рассмотрены два варианта решения. Первый реализует бинарный поиск вручную, позволяя лучше понять принцип работы алгоритма. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, благодаря чему код становится более компактным и удобным для использования.

Оба решения имеют одинаковую асимптотическую сложность и позволяют эффективно обрабатывать большое количество запросов, полностью удовлетворяя ограничениям задачи.
