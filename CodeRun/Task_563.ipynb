## Задача 563. Поход в горы

Хостел может одновременно вместить не более `s` человек. На ближайший период поступило `n` заявок, и для каждого туриста известно время его прибытия в хостел.

Необходимо установить одинаковое для всех туристов ограничение на максимальное время пребывания так, чтобы в любой момент времени в хостеле находилось не более `s` человек.

Требуется определить максимально возможное положительное время пребывания.

Если хостел будет переполнен при любом положительном времени пребывания, необходимо вывести `Impossible`.

Если ограничивать время пребывания вообще не требуется, необходимо вывести `INF`.

---

### Формат ввода

В первой строке записаны два целых числа:

- `n` — количество заявок;
- `s` — максимальное количество человек, которые могут одновременно находиться в хостеле.

В следующих `n` строках записано по одному целому числу `t_i` — время прибытия очередного туриста.

---

### Формат вывода

Выведите:

- максимальное положительное время, которое каждый турист может провести в хостеле;
- `Impossible`, если при любом положительном времени хостел окажется переполнен;
- `INF`, если устанавливать ограничение на время пребывания не требуется.

---

### Ограничения

- `1 ≤ n, s ≤ 2 · 10⁵`;
- `0 ≤ t_i ≤ 2 · 10⁹`;
- ограничение времени — 1 секунда;
- ограничение памяти — 64 МБ.

## 1) Решение с помощью бинарного поиска по ответу

Необходимо найти максимальное время пребывания туристов в хостеле, при котором одновременно в нём находится не более `s` человек.

Если некоторое время подходит, то любое меньшее время также будет подходить. Благодаря этому можно применить бинарный поиск по ответу.

---

### Алгоритм

1. Если количество туристов не превышает вместимость хостела, вывести `INF`.
2. Отсортировать времена прибытия.
3. Проверить возможность пребывания в течение одной единицы времени.
4. Если это невозможно, вывести `Impossible`.
5. Выполнить бинарный поиск по времени пребывания.
6. Вывести максимальное подходящее значение.

---

### Проверка (`check`)

Для проверки используется метод двух указателей.

Правый указатель последовательно рассматривает каждого нового туриста, а левый сдвигается до тех пор, пока самый ранний турист не покинет хостел.

```
while lst[left] + t <= lst[right]:
    left += 1
```

В каждый момент количество туристов в хостеле равно

```
right - left + 1
```

Если это значение становится больше `s`, выбранное время пребывания не подходит.

---

### Бинарный поиск

Минимальное возможное время равно `1`, а максимальная граница выбирается больше любого возможного ответа.

```
left = 1
right = lst[-1] - lst[0] + 1
```

Если функция `check` возвращает `True`, время пребывания можно увеличить.

```
left = mid
```

Иначе его необходимо уменьшить.

```
right = mid
```

После завершения поиска в переменной `left` находится максимальное допустимое время пребывания.

---

### Особенности

- перед поиском времена прибытия сортируются;
- проверка выполняется методом двух указателей;
- бинарный поиск позволяет быстро найти максимальное допустимое время.

---

### Сложность

Сортировка занимает `O(n log n)`.

Функция `check` работает за `O(n)`, а бинарный поиск выполняет `O(log T)` проверок, где `T` — диапазон возможных значений времени пребывания.

- **Время:** `O(n log n + n log T)`.
- **Дополнительная память:** `O(1)`.

````python
def check(n: int,  s: int, lst: list[int], t: int) -> bool:
    left = 0

    for right in range(n):
        while lst[left] + t <= lst[right]:
            left += 1

        if right - left + 1 > s:
            return False

    return True


def main():
    # n, s = map(int, input().split())
    # lst = [int(input()) for _ in range(n)]

    n, s = 8, 3
    lst = [3, 6, 4, 5, 0, 2, 7, 0]

    if n <= s:
        print('INF')
        return

    lst.sort()
    if not check(n, s, lst, 1):
        print('Impossible')
        return


    left = 1
    right = lst[-1] - lst[0] + 1

    while right - left > 1:
        mid = (left + right) // 2
        if check(n, s, lst, mid):
            left = mid
        else:
            right = mid

    print(left)


if __name__ == '__main__':
    main()
````

## 2) Решение с помощью модуля `bisect`

Данное решение использует ту же идею, что и предыдущее. Отличие заключается лишь в том, что вместо самостоятельной реализации бинарного поиска применяется функция `bisect_left` из стандартного модуля `bisect`.

---

### Алгоритм

1. Если количество туристов не превышает вместимость хостела, вывести `INF`.
2. Отсортировать времена прибытия.
3. Проверить возможность пребывания в течение одной единицы времени.
4. Если это невозможно, вывести `Impossible`.
5. С помощью `bisect_left` найти первое неподходящее время пребывания.
6. Вывести найденный ответ.

---

### Использование `bisect_left`

Функция `check` формирует монотонную последовательность.

```
True True True False False
```

Чтобы использовать `bisect_left`, результат проверки инвертируется.

```
False False False True True
```

Это позволяет найти первое значение, при котором время пребывания становится слишком большим.

```
answer = bisect_left(
    range(1, right + 1),
    True,
    key=lambda t: not check(n, s, lst, t)
)
```

Так как поиск выполняется по диапазону, начинающемуся с `1`, индекс первого неподходящего значения совпадает с максимальным допустимым временем пребывания.

---

### Особенности

- используется та же функция `check`;
- бинарный поиск выполняет функция `bisect_left`;
- код получается более компактным.

---

### Сложность

Сортировка занимает `O(n log n)`.

Функция `check` работает за `O(n)`, а `bisect_left` выполняет `O(log T)` проверок, где `T` — диапазон возможных значений времени пребывания.

- **Время:** `O(n log n + n log T)`.
- **Дополнительная память:** `O(1)`.

````python
from bisect import bisect_left

def check(n: int,  s: int, lst: list[int], t: int) -> bool:
    left = 0

    for right in range(n):
        while lst[left] + t <= lst[right]:
            left += 1

        if right - left + 1 > s:
            return False

    return True


def main():
    # n, s = map(int, input().split())
    # lst = [int(input()) for _ in range(n)]

    n, s = 8, 3
    lst = [3, 6, 4, 5, 0, 2, 7, 0]

    if n <= s:
        print('INF')
        return

    lst.sort()
    if not check(n, s, lst, 1):
        print('Impossible')
        return

    right = lst[-1] - lst[0] + 1

    answer = bisect_left(range(1, right + 1),
                         True,
                         key=lambda t: not check(n, s, lst, t))
    print(answer)

if __name__ == '__main__':
    main()
````

## Вывод

В данной задаче требовалось определить максимальное время пребывания туристов в хостеле, при котором одновременно в нём будет находиться не более `s` человек.

Для решения был использован **бинарный поиск по ответу**. Проверка каждого варианта выполнялась с помощью метода двух указателей, позволяющего за линейное время определить, не превышает ли количество одновременно находящихся в хостеле туристов допустимую вместимость.

Были рассмотрены два варианта реализации. Первый реализует бинарный поиск вручную, что позволяет лучше понять принцип его работы. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, благодаря чему код становится более компактным и лаконичным. Оба решения имеют одинаковую асимптотическую сложность и эффективно работают даже при максимальных ограничениях задачи.
