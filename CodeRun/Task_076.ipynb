## Задача 76. Приближённый двоичный поиск

Даны две последовательности чисел.
Первая последовательность содержит `N` чисел и отсортирована по неубыванию.
Для каждого числа из второй последовательности необходимо найти число из первой последовательности, которое находится к нему ближе всего.
Если ближайших чисел несколько, нужно вывести меньшее из них.

## Формат ввода
В первой строке содержатся два целых числа:
```
N K
```

где:
- `N` — количество элементов в первом массиве;
- `K` — количество элементов во втором массиве.

Во второй строке содержатся `N` чисел первого массива, отсортированных по неубыванию.
В третьей строке содержатся `K` чисел второго массива.

## Формат вывода
Для каждого числа из второго массива выведите в отдельной строке ближайшее к нему число из первого массива.
Если существуют два одинаково близких числа, выведите меньшее из них.

## Ограничения
- `0 < N, K < 100001`
- Каждое число по модулю не превосходит `2 · 10⁹`
- Ограничение времени — **1 секунда**.
- Ограничение памяти — **64 МБ**.

## 1) Приближённый двоичный поиск

Так как первый массив уже отсортирован, для каждого числа из второй последовательности можно использовать двоичный поиск.
Сначала проверяются два граничных случая:
- если искомое число не больше первого элемента массива, ответом будет первый элемент;
- если искомое число не меньше последнего элемента массива, ответом будет последний элемент.

В остальных случаях двоичный поиск сужает диапазон поиска до двух соседних элементов, между которыми находится искомое число.
После завершения поиска остаётся сравнить расстояние до этих двух элементов и выбрать ближайший. Если расстояния равны, выбирается меньший элемент, как требуется по условию задачи.

### Алгоритм

1. Проверить, находится ли искомое число левее или правее границ массива.
2. Выполнить двоичный поиск:
   - если найдено точное совпадение, вернуть его;
   - иначе постепенно сузить диапазон поиска до двух соседних элементов.
3. Сравнить расстояние до левого и правого элемента.
4. Вернуть ближайшее значение. При равных расстояниях вернуть левый элемент.
5. Повторить поиск для каждого числа второго массива.

### Особенности

- Используется классический двоичный поиск.
- После завершения поиска остаются два соседних кандидата на ответ.
- При одинаковом расстоянии до двух чисел выбирается меньшее.
- Для каждого запроса выполняется независимый двоичный поиск.
- Временная сложность одного поиска — **O(log N)**.
- Общая временная сложность — **O(K · log N)**.
- Дополнительная память — **O(1)**.

````python
def binary_search(arr: list[int], t: int) -> int:
    left = 0
    right = len(arr) - 1

    if t <= arr[left]:
        return arr[left]
    elif arr[right] <= t:
        return arr[right]

    while right - left > 1:
        m = (left + right) // 2
        if arr[m] == t:
            return arr[m]
        elif arr[m] < t:
            left = m
        else:
            right = m

    if t - arr[left] <= arr[right] - t:
        return arr[left]
    return arr[right]


def main():
    # n, k = map(int, input())
    # arr_n = list(map(int, input().split()))
    # arr_k = map(int, input().split())

    n, k = 5, 5
    arr_n = [1, 3, 5, 7, 9]
    arr_k = (2, 4, 8, 1, 6)

    for i in arr_k:
        print(binary_search(arr_n, i))


if __name__ == '__main__':
    main()
````

## 2) Приближённый поиск с использованием `bisect_left`

Модуль `bisect` содержит готовую реализацию двоичного поиска.

Функция `bisect_left()` возвращает позицию, в которую нужно вставить число, чтобы массив оставался отсортированным. Другими словами, она находит индекс первого элемента, который не меньше искомого числа.

После получения этой позиции остаётся рассмотреть два возможных кандидата:

- элемент слева;
- элемент справа.

Если искомое число оказалось меньше первого элемента массива или больше последнего, ответ определяется сразу. В остальных случаях выбирается ближайший из двух соседних элементов. Если расстояния одинаковы, выбирается меньший элемент.

### Алгоритм

1. Для каждого числа из второго массива вызвать `bisect_left()`.
2. Если полученный индекс равен `0`, вернуть первый элемент массива.
3. Если индекс равен длине массива, вернуть последний элемент.
4. Сравнить расстояния до элементов слева и справа.
5. Вернуть ближайший элемент. При равных расстояниях вернуть левый.

### Особенности

- Используется готовая реализация двоичного поиска из стандартной библиотеки Python.
- После поиска сравниваются только два соседних элемента.
- При одинаковом расстоянии выбирается меньшее число.
- Код получается значительно короче, чем при собственной реализации двоичного поиска.
- Временная сложность одного поиска — **O(log N)**.
- Общая временная сложность — **O(K · log N)**.
- Дополнительная память — **O(1)**.

````python
from bisect import bisect_left


def binary_search(arr: list[int], target: int) -> int:
    pos = bisect_left(arr, target)

    if pos == 0:
        return arr[0]

    if pos == len(arr):
        return arr[-1]

    if target - arr[pos - 1] <= arr[pos] - target:
        return arr[pos - 1]

    return arr[pos]


def main():
    # n, k = map(int, input())
    # arr_n = list(map(int, input().split()))
    # arr_k = map(int, input().split())

    n, k = 5, 5
    arr_n = [1, 3, 5, 7, 9]
    arr_k = (2, 4, 8, 1, 6)

    for i in arr_k:
        print(binary_search(arr_n, i))


if __name__ == '__main__':
    main()
````

## Вывод

В данной задаче были рассмотрены два способа поиска ближайшего элемента в отсортированном массиве.

В первом решении двоичный поиск реализован вручную. Такой подход позволяет лучше понять принцип работы алгоритма, обработку граничных случаев и выбор ближайшего элемента после завершения поиска.

Во втором решении используется функция `bisect_left()` из стандартной библиотеки Python. Она выполняет тот же двоичный поиск, благодаря чему код становится короче и проще, а его асимптотическая сложность остаётся такой же.

Оба решения имеют одинаковые характеристики:

- время одного поиска — **O(log N)**;
- общая временная сложность — **O(K · log N)**;
- дополнительная память — **O(1)**.

Если цель — изучить алгоритм двоичного поиска, рекомендуется реализовать его самостоятельно. В практических проектах на Python чаще используют `bisect_left()`, поскольку это готовое, надёжное и лаконичное решение.
