## Задача 245. Коллекционер Диего

Диего собирает наклейки с различными номерами. Некоторые наклейки в его коллекции могут иметь одинаковые номера, поэтому одинаковые наклейки считаются одним видом.

К Диего приходят коллекционеры. Каждый коллекционер сообщает номер наклейки, которую он уже нашёл.

Для каждого коллекционера необходимо определить, сколько различных наклеек с меньшими номерами есть у Диего.

---

### Формат ввода

В первой строке записано целое число `n` — количество наклеек у Диего.

Во второй строке записаны `n` целых чисел — номера наклеек.

В третьей строке записано целое число `k` — количество коллекционеров.

В четвёртой строке записаны `k` целых чисел — номера наклеек, найденных коллекционерами.

---

### Формат вывода

Для каждого коллекционера выведите количество различных наклеек Диего, номера которых строго меньше указанного коллекционером номера.

Каждый ответ необходимо вывести на отдельной строке.

---

### Ограничения

- `1 ≤ n, k ≤ 100 000`;
- номера наклеек не превышают `10⁹`;
- ограничение времени — 2 секунды;
- ограничение памяти — 256 МБ.

---

### Пример

**Ввод**

```
3
5 1 5
2
4 6
```

**Вывод**

```
1
2
```

После удаления повторяющихся номеров коллекция Диего имеет вид:

```
1 5
```

Для номера `4` существует одна наклейка с меньшим номером — `1`.

Для номера `6` существуют две наклейки с меньшими номерами — `1` и `5`.

## 1) Решение с помощью бинарного поиска

Так как одинаковые наклейки считаются одним видом, сначала необходимо удалить повторяющиеся номера и отсортировать коллекцию.

После этого для каждого запроса с помощью бинарного поиска определяется количество номеров, которые строго меньше заданного значения.

---

### Алгоритм

1. Удалить повторяющиеся номера наклеек.
2. Отсортировать полученный массив.
3. Для каждого запроса выполнить бинарный поиск.
4. Найти количество элементов, меньших заданного номера.
5. Вывести найденный результат.

---

### Подготовка данных

Перед обработкой запросов удаляются повторяющиеся номера и выполняется сортировка.

```
lst_n = sorted(set(lst_n))
```

После этого все номера располагаются в порядке возрастания, а каждый вид наклейки встречается только один раз.

---

### Бинарный поиск (`bin_search`)

Функция ищет индекс последнего элемента, не превышающего значение `t`.

```
arr[mid] <= t
```

Если условие выполняется, поиск продолжается в правой части массива.

```
left = mid
```

Иначе поиск продолжается слева.

```
right = mid - 1
```

После завершения поиска выполняется дополнительная проверка.

```
if left == 0 and arr[left] >= t:
    return 0
```

Если первый элемент массива не меньше `t`, значит подходящих элементов нет.

Во всех остальных случаях количество элементов, меньших или равных найденному, равно

```
left + 1
```

---

### Особенности

- перед поиском удаляются повторяющиеся номера;
- используется бинарный поиск для обработки каждого запроса;
- решение эффективно работает при большом количестве запросов.

---

### Сложность

Подготовка массива занимает `O(n + m log m)`, где `m` — количество различных наклеек.

Каждый запрос обрабатывается за `O(log m)`.

Итоговая сложность:

- **Время:** `O(n + m log m + k log m)`.

- **Дополнительная память:** `O(m)`.

````python
def bin_search(arr: list[int], t: int) -> int:
    left = 0
    right = len(arr) - 1
    while left < right:
        mid = (left + right + 1) // 2
        if arr[mid] <= t:
            left = mid
        else:
            right = mid - 1

    if left == 0 and arr[left] >= t:
        return 0
    return left + 1

def main():
    # n = int(input())
    # lst_n = list(map(int, input().split()))
    # k = int(input())
    # lst_k = map(int, input().split())

    n = 3
    lst_n = [5, 1, 5]
    k = 2
    lst_k = [4, 6]

    lst_n = sorted(set(lst_n))

    for i in lst_k:
        print(bin_search(lst_n, i))


main()
````

## 2) Решение с помощью модуля `bisect`

Вместо собственной реализации бинарного поиска можно воспользоваться функцией `bisect_left` из стандартного модуля `bisect`.

После удаления повторяющихся номеров и сортировки коллекции достаточно для каждого запроса определить позицию, на которую можно вставить число, не нарушая порядок элементов. Эта позиция и будет количеством наклеек с меньшими номерами.

---

### Алгоритм

1. Удалить повторяющиеся номера наклеек.
2. Отсортировать полученный массив.
3. Для каждого запроса вызвать функцию `bisect_left`.
4. Вывести полученный результат.

---

### Подготовка данных

Перед обработкой запросов удаляются повторяющиеся номера и выполняется сортировка.

```
lst_n = sorted(set(lst_n))
```

После этого каждый номер встречается только один раз, а все элементы располагаются в порядке возрастания.

---

### Поиск ответа

Для обработки каждого запроса используется функция

```
bisect_left(lst_n, number)
```

Она возвращает индекс первой позиции, в которую можно вставить число `number`, сохранив сортировку массива.

Все элементы, расположенные левее этой позиции, строго меньше искомого значения. Следовательно, найденный индекс одновременно является количеством различных наклеек, номера которых меньше `number`.

Например, для массива

```
[1, 50, 100]
```

и запроса

```
75
```

функция вернёт

```
2
```

поскольку числа `1` и `50` меньше `75`.

---

### Особенности

- используется готовая реализация бинарного поиска;
- код получается значительно компактнее;
- результат `bisect_left` сразу является ответом задачи;
- решение эффективно работает при большом количестве запросов.

---

### Сложность

Подготовка массива занимает `O(n + m log m)`, где `m` — количество различных наклеек.

Каждый запрос обрабатывается за `O(log m)`.

Итоговая сложность:

- **Время:** `O(n + m log m + k log m)`.
- **Дополнительная память:** `O(m)`.

````python
from bisect import bisect_left


def main():
    # n = int(input())
    # lst_n = list(map(int, input().split()))
    # k = int(input())
    # lst_k = map(int, input().split())

    n = 3
    lst_n = [5, 1, 5]
    k = 2
    lst_k = [4, 6]

    lst_n = sorted(set(lst_n))

    for number in lst_k:
        print(bisect_left(lst_n, number))


main()
````

## Вывод

В данной задаче требовалось для каждого запроса определить количество различных наклеек Диего, номера которых строго меньше заданного значения.

Для эффективного решения сначала были удалены повторяющиеся номера и выполнена сортировка коллекции. После этого каждый запрос обрабатывался с помощью бинарного поиска, что позволило находить ответ за логарифмическое время.

Были рассмотрены два варианта реализации. Первый реализует бинарный поиск вручную и позволяет подробно разобраться в принципе его работы. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, благодаря чему решение становится более компактным и лаконичным. Оба подхода имеют одинаковую асимптотическую сложность и эффективно справляются даже с очень большим количеством запросов.
