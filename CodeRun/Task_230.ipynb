## Задача 230. Провода

На складе находятся `n` отрезков провода различной длины.
Необходимо разрезать их на `k` одинаковых отрезков максимально возможной длины. Полученные отрезки должны иметь целую длину.
Оставшиеся части проводов можно не использовать.
Требуется определить максимальную длину одного отрезка.

---

### Формат ввода

В первой строке записаны два целых числа:

- `n` — количество исходных проводов;
- `k` — необходимое количество одинаковых отрезков.

В следующих `n` строках записано по одному целому числу — длины исходных проводов.

---

### Формат вывода

Выведите одно целое число — максимальную возможную длину каждого из `k` одинаковых отрезков.

Если получить необходимое количество отрезков положительной длины невозможно, выведите `0`.

---

### Ограничения

- ограничение времени — 1 секунда;
- ограничение памяти — 64 МБ.

---

### Пример

**Ввод**

```
4 11
802
743
457
539
```

**Вывод**

```
200
```

Из проводов можно получить не менее `11` отрезков длиной `200`:

```
802 // 200 = 4
743 // 200 = 3
457 // 200 = 2
539 // 200 = 2
```

Общее количество отрезков:

```
4 + 3 + 2 + 2 = 11
```

Получить `11` отрезков длиной `201` уже невозможно, поэтому максимальная подходящая длина равна `200`.

## 1) Решение с помощью бинарного поиска по ответу

Требуется найти **максимальную** длину одинаковых отрезков, на которые можно разрезать провода.

Если для длины `x` удаётся получить не менее `k` отрезков, то любая меньшая длина также подойдёт. Это позволяет использовать **бинарный поиск по ответу**.

---

### Алгоритм

1. Задать диапазон возможных ответов.
2. Найти середину диапазона.
3. Проверить, можно ли получить не менее `k` отрезков выбранной длины.
4. Если можно — попробовать увеличить длину.
5. Иначе уменьшить длину.
6. Повторять, пока не будет найден максимальный возможный ответ.

---

### Проверка (`check`)

Для каждого провода вычисляется количество отрезков длины `x`, которое можно из него получить.

```
i // x
```

Затем количество отрезков суммируется.

```
sum(i // x for i in arr)
```

Если полученное количество не меньше `k`, значит длина `x` подходит.

```
sum(i // x for i in arr) >= k
```

---

### Бинарный поиск

Левая граница поиска равна `0`, а правая — длине самого длинного провода.

```
left = 0
right = max(arr)
```

Во время поиска вычисляется середина диапазона.

```
mid = (left + right + 1) // 2
```

Добавление единицы позволяет избежать зацикливания, когда необходимо найти **максимальный** подходящий ответ.

Если проверка успешна, левая граница сдвигается вправо.

```
left = mid
```

Иначе уменьшается правая граница.

```
right = mid - 1
```

После завершения поиска переменная `left` содержит максимальную возможную длину одинаковых отрезков.

---

### Особенности

- используется бинарный поиск по ответу;
- каждая проверка выполняется за `O(n)`;
- решение эффективно работает даже при больших размерах входных данных.

---

### Сложность

Каждая проверка проходит по всем проводам и выполняется за `O(n)`, а бинарный поиск делает `O(log L)` проверок, где `L` — длина самого длинного провода.

- **Время:** `O(n log L)`.
- **Дополнительная память:** `O(1)`.

````python
def check(arr: list[int], k: int, x: int) -> bool:
    return sum([i // x for i in arr]) >= k

def main():
    # n, k = map(int, input().split())
    # arr = [int(input()) for _ in range(n)]

    n, k = 4, 11
    arr = [802, 743, 457, 539]

    left = 0
    right = max(arr)
    while left < right:
        mid = (left + right + 1) // 2
        if check(arr, k, mid):
            left = mid
        else:
            right = mid - 1
    print(left)

main()
````

## 2) Решение с помощью модуля `bisect`

Вместо собственной реализации бинарного поиска можно воспользоваться функцией `bisect_left` из стандартного модуля `bisect`.

Поскольку необходимо найти **максимальную** подходящую длину, поиск выполняется по первой длине, которая уже не удовлетворяет условию. После этого её индекс и будет являться искомым ответом.

---

### Алгоритм

1. Задать диапазон возможных длин отрезков.
2. Для каждой длины выполнять проверку с помощью функции `check`.
3. Найти первую длину, для которой проверка возвращает `False`.
4. Вывести найденное значение.

---

### Поиск ответа

Для поиска используется функция `bisect_left`.

```
answer = bisect_left(
    range(1, right + 2),
    True,
    key=lambda x: not check(arr, k, x)
)
```

Функция `check` возвращает `True`, если из проводов можно получить не менее `k` отрезков длины `x`.

Так как `bisect_left` ищет первое значение `True`, результат проверки инвертируется с помощью оператора `not`.

```
not check(arr, k, x)
```

После этого последовательность приобретает вид:

```
False False ... False True True ...
```

Функция `bisect_left` находит первое значение `True`, которое соответствует минимальной неподходящей длине. Поскольку поиск выполняется по диапазону, начинающемуся с `1`, полученный индекс совпадает с максимальной подходящей длиной.

---

### Особенности

- используется готовая реализация бинарного поиска;
- функция `check` остаётся без изменений;
- решение имеет ту же асимптотическую сложность, что и ручная реализация.

---

### Сложность

Каждая проверка выполняется за `O(n)`, а бинарный поиск делает `O(log L)` проверок, где `L` — длина самого длинного провода.

- **Время:** `O(n log L)`.
- **Дополнительная память:** `O(1)`.

````python
from bisect import bisect_left


def check(arr: list[int], k: int, x: int) -> bool:
    return sum(i // x for i in arr) >= k


def main():
    # n, k = map(int, input().split())
    # arr = [int(input()) for _ in range(n)]

    n, k = 4, 11
    arr = [802, 743, 457, 539]

    right = max(arr)

    answer = bisect_left(range(1, right + 2),
                         True,
                         key=lambda x: not check(arr, k, x))

    print(answer)


main()
````

## Вывод

В данной задаче требовалось определить максимальную длину одинаковых отрезков, которые можно получить из имеющихся проводов.

Для решения был использован **бинарный поиск по ответу**, позволяющий найти максимальную подходящую длину без перебора всех возможных вариантов. Каждая проверка сводится к подсчёту количества отрезков, которые можно получить при выбранной длине.

Были рассмотрены два варианта реализации. Первый реализует бинарный поиск вручную и наглядно демонстрирует принцип работы алгоритма. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, позволяя выполнить тот же поиск с помощью готовой реализации. Оба решения имеют одинаковую асимптотическую сложность и эффективно работают даже при больших размерах входных данных.
