## Задача 79. Субботник

В классе учатся `N` человек. Классный руководитель должен направить на субботник `R` бригад по `C` человек в каждой.

Все бригады будут заниматься переноской брёвен. Каждое бревно одновременно несут все участники одной бригады. Чем меньше различается рост участников бригады, тем удобнее им работать вместе.

**Числом неудобства бригады** называется разность между ростом самого высокого и самого низкого участника этой бригады.

Если в бригаде находится только один человек, число неудобства равно `0`.

Необходимо сформировать `R` бригад по `C` человек так, чтобы максимальное число неудобства среди всех сформированных бригад было как можно меньше.

Требуется определить минимально возможное значение этого максимального неудобства.

---

## Пример

Пусть в классе учатся `8` человек со следующим ростом:

```
170, 205, 225, 190, 260, 130, 225, 160
```

Необходимо сформировать `2` бригады по `3` человека.

Один из оптимальных вариантов:

```
1-я бригада: 205, 225, 225
2-я бригада: 160, 170, 190
```

Число неудобства первой бригады:

```text
225 - 205 = 20
```

Число неудобства второй бригады:

```
190 - 160 = 30
```

Максимальное число неудобства среди сформированных бригад равно:

```
max(20, 30) = 30
```

Получить значение меньше `30` невозможно, поэтому ответ равен `30`.

---

## Формат ввода

В первой строке вводятся три натуральных числа:

```
N R C
```

где:
- `N` — количество учеников в классе;
- `R` — количество бригад;
- `C` — количество человек в каждой бригаде.

Далее вводятся `N` целых чисел — рост каждого ученика.
Рост каждого ученика является натуральным числом и не превышает `10⁹`.

---

## Формат вывода

Выведите одно целое число — минимально возможное значение максимального числа неудобства среди сформированных бригад.

---

1 ≤ height[i] ≤ 1 000 000 000
```

```

## 1) Решение с помощью бинарного поиска по ответу

Так как требуется **минимизировать максимальное число неудобства**, напрямую подобрать ответ достаточно сложно. Однако можно заметить важную особенность: если некоторое значение неудобства `x` позволяет сформировать необходимые бригады, то любое большее значение также подойдёт. Это означает, что ответ обладает свойством монотонности, а значит задачу можно решить с помощью **бинарного поиска по ответу**.

Перед началом поиска отсортируем рост всех учеников. После сортировки наиболее подходящие кандидаты для одной бригады будут располагаться рядом друг с другом, поэтому достаточно рассматривать только непрерывные группы из `C` человек.

---

### Алгоритм

1. Отсортировать массив ростов.
2. Выполнить бинарный поиск по возможному значению максимального неудобства.
3. Для каждого предполагаемого значения `x` проверить, можно ли сформировать `R` бригад.
4. Если это возможно, попробовать уменьшить значение `x`.
5. Иначе увеличить границу поиска.
6. После завершения бинарного поиска найденное значение будет минимально возможным ответом.

---

### Проверка возможности (`check`)

Функция `check` последовательно просматривает отсортированный массив и пытается жадно сформировать бригады.
Для каждого возможного начала бригады рассматриваются подряд идущие `C` человек.
Если разность между максимальным и минимальным ростом этой группы не превышает `x`, то группа считается подходящей:

```
arr[i + c - 1] - arr[i] <= x
```

После этого:
- найденная бригада фиксируется;
- указатель сразу переносится за её пределы;
- количество оставшихся бригад уменьшается на единицу.

Если текущая группа не подходит, начало окна сдвигается на одного человека вправо, после чего выполняется следующая проверка.

Как только удаётся сформировать все `R` бригад, функция сразу возвращает `True`. Если после просмотра всего массива нужное количество бригад собрать не удалось, возвращается `False`.

Жадный подход является оптимальным, поскольку после сортировки всегда выгодно брать первую подходящую группу: она использует минимально возможный диапазон ростов и не ухудшает возможность сформировать остальные бригады.

---

### Особенности

- используется бинарный поиск по ответу;
- проверка выполняется жадным алгоритмом;
- сортировка позволяет рассматривать только соседние элементы;
- каждая проверка имеет линейную сложность.

---

### Сложность

- Сортировка массива: `O(n log n)`.
- Каждая проверка выполняется за `O(n)`.
- Бинарный поиск выполняет такую проверку несколько раз.

Итоговая сложность алгоритма:

- **Время:** `O(n log n)`.
- **Дополнительная память:** `O(1)` (без учёта сортировки).

````python
def check(arr: list[int], n: int, r: int, c: int, x: int) -> bool:
    i = 0
    while i + c <= n:
        if arr[i + c - 1] - arr[i] <= x:
            r -= 1
            if r == 0:
                return True
            i += c
        else:
            i += 1
    return False


def main():
    # n, r, c = map(int, input().split())
    # array = [int(input()) for _ in range(n)]
    n, r, c = 8, 2, 3
    array = [170, 205, 225, 190, 260, 130, 225, 160]

    array.sort()

    left = 0
    right = array[-1] - array[0]

    while left < right:
        mid = (left + right) // 2
        if check(array, n, r, c, mid):
            right = mid
        else:
            left = mid + 1

    print(right)

main()
````

## 2) Решение с помощью модуля `bisect`

В предыдущем решении бинарный поиск по ответу был реализован вручную. В Python его также можно выполнить с помощью функции `bisect_left` из стандартного модуля `bisect`.

Основная идея решения не меняется. По-прежнему используется жадная проверка (`check`), которая определяет, можно ли сформировать необходимое количество бригад при заданном максимальном неудобстве. Отличие заключается лишь в том, что поиск минимального подходящего ответа выполняет уже готовая функция `bisect_left`.

---

### Алгоритм

1. Считать рост всех учеников.
2. Отсортировать массив по возрастанию.
3. Определить диапазон возможных значений максимального неудобства.
4. С помощью `bisect_left` найти минимальное значение, при котором функция `check` возвращает `True`.
5. Вывести найденный ответ.

---

### Диапазон возможных ответов

Минимальное значение неудобства равно `0`, а максимальное не может превышать разность между самым высоким и самым низким учениками.

Поэтому создаётся диапазон:

```python
possible_answers = range(array[-1] - array[0] + 1)
```

Объект `range` не создаёт список всех чисел в памяти, а лишь хранит границы диапазона, поэтому практически не расходует дополнительную память.

---

### Поиск ответа с помощью `bisect_left`

Для каждого возможного значения неудобства вызывается функция `check`.

Она определяет, можно ли сформировать требуемое количество бригад, если максимальная разница в росте внутри каждой из них не превышает выбранного значения.

Результаты работы функции можно представить в виде последовательности:

```
False, False, False, ..., True, True, True
```

Пока значение неудобства слишком маленькое, сформировать нужное количество бригад невозможно, поэтому функция возвращает `False`.

Начиная с некоторого значения это становится возможным, поэтому результат меняется на `True`.

Именно поэтому вторым аргументом функции `bisect_left` передаётся значение `True`.

```
answer = bisect_left(
    possible_answers,
    True,
    key=lambda x: check(array, n, r, c, x)
)
```

Функция последовательно применяет `check` к рассматриваемым значениям диапазона и ищет **первое значение**, для которого результат становится равен `True`.

Так как диапазон `possible_answers` начинается с нуля, найденный индекс одновременно является минимальным возможным значением максимального неудобства.

---

### Функция `check`

Функция проверки полностью совпадает с предыдущим решением.

Она последовательно просматривает отсортированный массив и с помощью жадного алгоритма пытается сформировать `R` бригад по `C` человек.

Если разница между первым и последним учеником очередной группы не превышает заданное значение неудобства, группа считается сформированной, после чего указатель сразу переходит к следующей возможной группе.

Если удаётся сформировать необходимое количество бригад, функция возвращает `True`, иначе — `False`.

---

### Особенности

- используется готовая реализация бинарного поиска из стандартной библиотеки;
- ручной бинарный поиск заменяется одной функцией `bisect_left`;
- жадная проверка остаётся без изменений;
- объект `range` не хранит все возможные ответы в памяти;
- решение требует Python 3.10 или новее, где `bisect_left` поддерживает параметр `key`.

---

### Сложность

- Сортировка массива: `O(n log n)`.
- Каждая проверка выполняется за `O(n)`.
- Бинарный поиск вызывает функцию проверки `O(log H)` раз, где `H` — диапазон возможных ответов.

Итоговая сложность:

- **Время:** `O(n log n + n log H)`.
- **Дополнительная память:** `O(1)` (без учёта сортировки).

````python
from bisect import bisect_left


def check(arr: list[int], n: int, r: int, c: int, x: int) -> bool:
    i = 0
    while i + c <= n:
        if arr[i + c - 1] - arr[i] <= x:
            r -= 1
            if r == 0:
                return True
            i += c
        else:
            i += 1
    return False


def main():
    # n, r, c = map(int, input().split())
    # array = [int(input()) for _ in range(n)]
    n, r, c = 8, 2, 3
    array = [170, 205, 225, 190, 260, 130, 225, 160]

    array.sort()

    possible_answers = range(array[-1] - array[0] + 1)

    answer = bisect_left(possible_answers,
                         True,
                         key=lambda x: check(array, n, r, c, x))

    print(answer)

main()
````

## Вывод

В этой задаче требовалось минимизировать максимальную разницу в росте учеников внутри сформированных бригад.

Для решения были использованы сразу два подхода:

1. **Жадный алгоритм** — для проверки, можно ли сформировать нужное количество бригад при заданном значении неудобства.
2. **Бинарный поиск по ответу** — для нахождения минимального подходящего значения неудобства.

Также были рассмотрены две реализации бинарного поиска: ручная и с помощью функции `bisect_left` из модуля `bisect`. Оба варианта используют одну и ту же монотонность функции проверки и имеют одинаковую асимптотическую сложность.

Таким образом, задача хорошо демонстрирует сочетание сортировки, жадного алгоритма и бинарного поиска по ответу.
