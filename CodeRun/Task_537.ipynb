## Задача 537. Покрытие K отрезками

Даны `n` точек на прямой. Необходимо покрыть все точки не более чем `k` отрезками одинаковой длины `l`.
Точки, расположенные на границах отрезка, также считаются покрытыми.
Требуется определить минимальную длину `l`, при которой все заданные точки можно покрыть `k` отрезками.

---

### Формат ввода
В первой строке записаны два целых числа:

- `n` — количество точек;
- `k` — максимальное количество отрезков.

Во второй строке записаны `n` целых чисел `x_i` — координаты точек на прямой.

---

### Формат вывода

Выведите одно целое число — минимальную длину отрезков, которой достаточно, чтобы покрыть все точки не более чем `k` отрезками.

---

### Ограничения

- `1 ≤ n ≤ 100 000`;
- `1 ≤ k ≤ n`;
- `|x_i| ≤ 10⁹`;
- ограничение времени — 1 секунда;
- ограничение памяти — 64 МБ.

---

### Пример

**Ввод**
```
6 2
1 2 3 9 8 7
```

**Вывод**
```
2
```

После сортировки координаты точек имеют вид:
```
1 2 3 7 8 9
```

Первые три точки можно покрыть отрезком длины `2`:
```
[1, 3]
```

Оставшиеся три точки можно покрыть вторым отрезком такой же длины:
```
[7, 9]
```

Таким образом, двух отрезков длины `2` достаточно, чтобы покрыть все точки.
Использовать меньшую длину не получится, поэтому ответ равен `2`.

## 1) Решение с помощью бинарного поиска по ответу

Требуется найти минимальную длину отрезков, которой достаточно, чтобы покрыть все точки не более чем `k` отрезками.

Заметим, что если некоторой длины достаточно для покрытия всех точек, то любой большей длины также будет достаточно. Значит, ответ обладает свойством монотонности, поэтому его можно найти с помощью бинарного поиска.

---

### Алгоритм

1. Отсортировать координаты точек.
2. Выполнять бинарный поиск по возможной длине отрезка.
3. Для каждой длины запускать функцию `check`.
4. Если покрыть все точки удалось, попробовать уменьшить длину.
5. Иначе увеличить длину.
6. После завершения поиска вывести минимальную подходящую длину.

---

### Проверка (`check`)

Функция проверяет, можно ли покрыть все точки не более чем `k` отрезками длины `l`.

Проверка выполняется жадным алгоритмом.

Берётся первая ещё не покрытая точка, после чего размещается отрезок так, чтобы его левый конец находился в этой точке.

```
end = lst[i] + l
```

Таким образом, текущий отрезок покрывает все точки с координатами, не превышающими

```
end
```

Пока очередная точка находится внутри этого отрезка, указатель сдвигается вправо.

```
while i < n and lst[i] <= end:
    i += 1
```

После этого строится следующий отрезок, начиная с первой непокрытой точки.

Если количество использованных отрезков превысило `k`, дальнейшая проверка уже не имеет смысла.

```
if segments > k:
    return False
```

Если удалось обработать все точки, функция возвращает

```
True
```

---

### Бинарный поиск

После сортировки точек минимальная возможная длина равна `0`, а максимальная — расстоянию между крайними точками.

```
left = 0
right = lst[-1] - lst[0]
```

Во время бинарного поиска проверяется средняя длина.

```
mid = (left + right) // 2
```

Если длины достаточно для покрытия всех точек,

```
check(...)
```

возвращает `True`, поэтому можно попробовать найти ещё меньшую длину.

```
right = mid
```

Иначе длину необходимо увеличить.

```
left = mid
```

После завершения поиска переменная

```
right
```

содержит минимальную подходящую длину.

---

### Особенности

- перед поиском координаты сортируются;
- проверка выполняется жадным алгоритмом;
- каждая точка просматривается не более одного раза;
- бинарный поиск позволяет быстро найти минимальную возможную длину.

---

### Сложность

Сортировка координат занимает `O(n log n)`.

Функция `check` работает за `O(n)`, а бинарный поиск выполняет `O(log D)` проверок, где `D` — расстояние между крайними точками.

Итоговая сложность:

- **Время:** `O(n log n + n log D)`.
- **Дополнительная память:** `O(1)`.

````python
def check(n: int, k: int, lst: list[int], l: int) -> bool:
    segments = 0
    i = 0

    while i < n:
        segments += 1
        end = lst[i] + l

        while i < n and lst[i] <= end:
            i += 1

        if segments > k:
            return False
    return True


def main():
    # n, k = map(int, input().split())
    # lst = list(map(int, input().split()))

    n, k = 6, 2
    lst = [1, 2, 3, 9, 8, 7]
    lst.sort()

    left = 0
    right = lst[-1] - lst[0]
    while right - left > 1:
        mid = (left + right) // 2
        if check(n, k, lst, mid):
            right = mid
        else:
            left = mid
    print(right)


if __name__ == '__main__':
    main()
````

## 2) Решение с помощью модуля `bisect`

Вместо собственной реализации бинарного поиска можно воспользоваться функцией `bisect_left` из стандартного модуля `bisect`.

Так как функция `check` обладает свойством монотонности, можно найти минимальную подходящую длину отрезка с помощью готовой реализации бинарного поиска.

---

### Алгоритм

1. Отсортировать координаты точек.
2. Задать диапазон возможных длин от `0` до максимального расстояния между крайними точками.
3. Для каждой длины выполнять проверку с помощью функции `check`.
4. Найти первую длину, для которой проверка возвращает `True`.
5. Вывести найденную длину.

---

### Проверка (`check`)

Функция полностью совпадает с предыдущим решением.

Она определяет, можно ли покрыть все точки не более чем `k` отрезками длины `l`, используя жадный алгоритм.

Если длины достаточно, функция возвращает

```
True
```

иначе —

```
False
```

---

### Поиск ответа

Для поиска минимальной подходящей длины используется функция

```
bisect_left(
    range(right + 1),
    True,
    key=lambda x: check(n, k, lst, x)
)
```

Для каждой возможной длины вызывается функция

```
check(n, k, lst, x)
```

которая формирует последовательность вида

```
False False ... False True True ...
```

Все недостаточные длины дают значение `False`, а начиная с минимальной подходящей длины функция начинает возвращать `True`.

Функция `bisect_left` находит первое значение `True`, которое и соответствует минимальной длине отрезков, достаточной для покрытия всех точек.

Поскольку поиск выполняется по диапазону

```
range(right + 1)
```

его индексы полностью совпадают со значениями длин, поэтому результат `bisect_left` сразу является ответом задачи.

---

### Особенности

- используется готовая реализация бинарного поиска;
- функция `check` остаётся без изменений;
- решение получается компактнее ручной реализации;
- время работы полностью совпадает с первым способом.

---

### Сложность

Сортировка координат занимает `O(n log n)`.

Функция `check` работает за `O(n)`, а бинарный поиск выполняет `O(log D)` проверок, где `D` — расстояние между крайними точками.

Итоговая сложность:

- **Время:** `O(n log n + n log D)`.
- **Дополнительная память:** `O(1)`.

````python
from bisect import bisect_left


def check(n: int, k: int, lst: list[int], l: int) -> bool:
    segments = 0
    i = 0

    while i < n:
        segments += 1
        end = lst[i] + l

        while i < n and lst[i] <= end:
            i += 1

        if segments > k:
            return False

    return True


def main():
    # n, k = map(int, input().split())
    # lst = list(map(int, input().split()))

    n, k = 6, 2
    lst = [1, 2, 3, 9, 8, 7]

    lst.sort()

    right = lst[-1] - lst[0]

    answer = bisect_left(range(right + 1),
                         True,
                         key=lambda x: check(n, k, lst, x))

    print(answer)


if __name__ == '__main__':
    main()
````

Для решения был использован **бинарный поиск по ответу**. Для каждой возможной длины выполнялась проверка с помощью жадного алгоритма, который последовательно покрывал максимальное количество точек каждым отрезком. Благодаря монотонности условия удалось эффективно найти минимальную подходящую длину.

Были рассмотрены два варианта реализации. Первый реализует бинарный поиск вручную и позволяет лучше понять принцип его работы. Второй использует функцию `bisect_left` из стандартного модуля `bisect`, благодаря чему код становится более компактным и лаконичным. Оба решения имеют одинаковую асимптотическую сложность и эффективно работают даже при максимальных ограничениях задачи.
