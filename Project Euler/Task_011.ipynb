## 1) Перебор всех направлений

В данном решении последовательно рассматривается каждая ячейка таблицы как возможное начало последовательности из четырёх чисел.

Для каждой позиции проверяются четыре направления:

- вправо;
- вниз;
- по диагонали вправо вниз;
- по диагонали влево вниз.

Проверять противоположные направления отдельно не требуется, поскольку те же последовательности уже будут рассмотрены с другой начальной позиции.

### Алгоритм

1. Определяется количество строк `r` и столбцов `c`.
2. Последовательно перебираются все элементы таблицы.
3. Если справа достаточно элементов, вычисляется произведение четырёх чисел по горизонтали.
4. Если снизу и справа достаточно элементов, проверяется диагональ вправо вниз.
5. Если снизу достаточно элементов, вычисляется произведение по вертикали.
6. Если снизу и слева достаточно элементов, проверяется диагональ влево вниз.
7. Каждое найденное произведение сравнивается с текущим максимумом.
8. После завершения перебора возвращается максимальное значение.

### Проверка границ

Перед вычислением произведения проверяется, помещается ли последовательность внутри таблицы.

Для движения вправо используется условие:

```python
j < c - 3
```

Для движения вниз:

```python
i < r - 3
```

Для диагонали влево вниз дополнительно проверяется:

```python
j > 2
```

Эти условия предотвращают выход индексов за границы таблицы.

### Особенности

- Каждая ячейка рассматривается как возможная начальная точка.
- Проверяются все необходимые направления.
- Противоположные направления отдельно не обрабатываются.
- Дополнительные структуры данных не используются.
- Временная сложность составляет `O(r · c)`.
- Пространственная сложность составляет `O(1)`.

### Вывод

Последовательный обход таблицы позволяет проверить все возможные расположения четырёх соседних чисел. Благодаря проверке только четырёх направлений и контролю границ решение остаётся компактным и эффективным.

````python
def get_result_1(nums: list[list[int]]) -> int:
    m = 0
    r, c = len(nums), len(nums[0])
    for i in range(r):
        for j in range(c):
            if j < c - 3:
                m = max(m, nums[i][j] * nums[i][j + 1] * nums[i][j + 2] * nums[i][j + 3])
                if i < r - 3:
                    m = max(m, nums[i][j] * nums[i + 1][j + 1] * nums[i + 2][j + 2] * nums[i + 3][j + 3])

            if i < r - 3:
                m = max(m, nums[i][j] * nums[i + 1][j] * nums[i + 2][j] * nums[i + 3][j])
                if j > 2:
                    m = max(m, nums[i][j] * nums[i + 1][j - 1] * nums[i + 2][j - 2] * nums[i + 3][j - 3])                      
    return m



start = perf_counter()
print(get_result_1(d))
end = perf_counter()

print(f'{end - start:.5f}')
````

## 2) Использование направлений

В данном решении используется тот же алгоритм, однако вместо отдельных проверок для каждого направления применяется единый список направлений.

Каждое направление задаётся парой смещений `(dr, dc)`, после чего произведение четырёх чисел вычисляется одинаковым способом независимо от ориентации.

### Отличия от предыдущего решения

- Все направления объединены в одну коллекцию.
- Проверка границ выполняется по единому правилу.
- Для вычисления произведения используется функция `prod()`.
- Код становится более компактным и легче расширяется.
- Алгоритм и временная сложность остаются без изменений.

### Вывод

Использование списка направлений позволяет избавиться от повторяющегося кода и сделать решение более универсальным.

````python
from math import prod


def get_result_2(nums: list[list[int]]) -> int:
    rows = len(nums)
    cols = len(nums[0])

    directions = (
        (0, 1),   # вправо
        (1, 0),   # вниз
        (1, 1),   # диагональ вправо вниз
        (1, -1),  # диагональ влево вниз
    )

    mx = 0
    for row in range(rows):
        for col in range(cols):
            for dr, dc in directions:
                if 0 <= row + 3 * dr < rows and 0 <= col + 3 * dc < cols:
                    mx = max(mx, prod(nums[row + step * dr][col + step * dc] for step in range(4)))

    return mx


start = perf_counter()
print(get_result_2(d))
end = perf_counter()

print(f'{end - start:.5f}')
````

## Сравнение времени выполнения

Для сравнения производительности используется функция `timeit()`, предназначенная специально для измерения времени выполнения небольших участков кода.

Каждый алгоритм запускается **1000 раз**, после чего общее время делится на количество запусков. Таким образом получается **среднее время выполнения одного запуска**.

````python
from timeit import timeit

funcs = (
    get_result_1,
    get_result_2,
)

for i, func in enumerate(funcs, start=1):
    tm = timeit(lambda: func(d), number=1000) / 1000
    print(f'Алгоритм {i}: {tm:.7f}')
````

## Заключение

В задаче были рассмотрены два подхода к поиску максимального произведения четырёх последовательных чисел в таблице.

Первый алгоритм проверяет каждое направление отдельным условием и напрямую перемножает четыре нужных элемента. Такой код содержит повторяющиеся выражения, но выполняет минимальное количество дополнительных операций.

Во втором решении направления объединены в одну коллекцию, а произведение вычисляется универсально с помощью функции `prod()` и генераторного выражения. Благодаря этому код становится компактнее, понятнее и легче расширяется, однако появляются дополнительные накладные расходы.

Среднее время выполнения алгоритмов:

- **Алгоритм 1:** `0.0003179` с
- **Алгоритм 2:** `0.0012966` с

Второй алгоритм оказался примерно в **4 раза медленнее** первого по нескольким причинам:

- для каждой ячейки дополнительно перебирается коллекция направлений;
- многократно вычисляются координаты `row + step * dr` и `col + step * dc`;
- для каждого подходящего направления создаётся генераторное выражение;
- вызывается универсальная функция `prod()`;
- выполняется внутренний цикл `range(4)` вместо четырёх прямых обращений к элементам таблицы.

Оба решения имеют одинаковую временную сложность `O(r · c)`, поскольку число направлений и длина последовательности являются константами. Разница возникает не из-за асимптотики, а из-за стоимости дополнительных операций внутри каждого шага.

Таким образом, первое решение лучше по скорости, а второе — по универсальности и удобству сопровождения. Для фиксированной задачи с четырьмя заранее известными направлениями прямые вычисления оказываются эффективнее обобщённой реализации.
