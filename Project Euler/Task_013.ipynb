## Задача 13

В файле `Task_13_file.txt` записаны сто чисел, каждое из которых содержит `50` цифр.

Необходимо вычислить сумму всех чисел и определить **первые десять цифр** полученного результата.

Ответ: 5537376230

## 1) Потоковое чтение и суммирование чисел

В данном решении числа считываются из файла последовательно и сразу преобразуются в тип `int`.

Функция `sum()` вычисляет сумму всех чисел, поступающих из `map()`, поэтому создавать отдельный список не требуется.

После этого полученная сумма преобразуется в строку, из которой берутся первые десять цифр.

### Алгоритм

1. Открывается файл с исходными числами.
2. Каждая строка преобразуется в целое число.
3. Все числа суммируются с помощью функции `sum()`.
4. Полученная сумма преобразуется в строку.
5. Возвращаются первые десять цифр результата.

### Особенности

- Используется встроенная поддержка целых чисел произвольной длины.
- Файл обрабатывается последовательно.
- Отдельный список чисел не создаётся.
- Временная сложность составляет `O(n)`.
- Дополнительная память относительно количества строк составляет `O(1)`.

### Вывод

Встроенные возможности Python позволяют компактно вычислить сумму больших чисел. Последовательная обработка файла не требует хранения всех значений в отдельном списке и делает решение простым и экономным по памяти.

```python
from time import perf_counter # измерение времени выполнения алгоритма


def get_result(file_path: str) -> str:
    with open(file_path) as file:
        sm = sum(map(int, file))

    return str(sm)[:10]
    

start = perf_counter()
print(get_result('../Sources/Project_Euler_013_file.txt'))
end = perf_counter()

print(f'{end - start:.5f}')
```

## Среднее время выполнения

Для измерения производительности используется функция `timeit()`, предназначенная для точного замера времени выполнения небольших участков кода.

Алгоритм запускается **1000 раз**, после чего вычисляется **среднее время выполнения одного запуска**.

```python
from timeit import timeit

tm = timeit(lambda: get_result('../Sources/Project_Euler_013_file.txt'), number=1000) / 1000
print(f'Среднее время: {tm:.5f}')
```
