### Задача 1 

Если выписать все натуральные числа меньше 10, кратные 3 или 5, то получим 3, 5, 6 и 9. Сумма этих чисел равна 23.

Найдите сумму всех чисел меньше 1000, кратных 3 или 5.


Ответ: 233168

### Импорт необходимых модулей

В этом блоке подключаем несколько стандартных библиотек Python:

- **time.perf_counter** — используется для точного измерения времени выполнения кода 
- **statistics.mean** — позволяет вычислять среднее значение, пригодится для анализа результатов 
- **matplotlib.pyplot** — библиотека для визуализации данных, с её помощью построим графики производительности

```python
from time import perf_counter

from statistics import mean
import matplotlib.pyplot as plt
```

### Решение задачи и измерение времени выполнения

1. С помощью цикла **for** перебираем все числа от 1 до 999
2. Проверяем, делится ли каждое число на 3 или 5
3. Если условие выполняется — прибавляем число к сумме **c**
4. После завершения цикла выводим итоговую сумму и время, затраченное на выполнение кода, измеренное с помощью **perf_counter()**

```python
start = perf_counter()

c = 0
for i in range(1, 1_000):
    if i % 3 == 0 or i % 5 == 0:
        c += i
print(c)  

end = perf_counter()

print(f'{end - start:.5f}')
```

### Решение через функцию

1. Функция принимает число **n** и считает сумму всех чисел меньше **n**, кратных 3 или 5
2. Логика такая же, как в предыдущем блоке: перебор чисел с условием и накопление суммы
3. Вызов функции для **n = 1000** выводит результат, а **perf_counter()** измеряет время её выполнения

Использование функции упрощает повторное применение к разным значениям **n**

```python
def get_result_2(n):
    c = 0
    for i in range(1, n):
        if i % 3 == 0 or i % 5 == 0:
            c += i
    return c


start = perf_counter()
print(get_result_2(1_000))
end = perf_counter()

print(f'{end - start:.5f}')
```

### Математическое решение через арифметическую прогрессию

В этом блоке используется более эффективный подход:

1. **get_number(a, b, n)** — вычисляет сумму всех чисел от **a** до **b**, кратных **n**, используя формулу суммы арифметической прогрессии:  
   - Находим первое число, кратное **n**, в диапазоне
   - Находим последнее число, кратное **n**, в диапазоне 
   - Считаем количество членов и сумму прогрессии
2. **get_result_3(a, b)** — суммирует числа, кратные 3 и 5, но вычитает числа, кратные 15, чтобы избежать двойного учёта
3. Такой подход значительно быстрее, особенно для больших диапазонов, так как не требует перебора всех чисел

```python
def get_number(a, b, n):
    ost_a = a % n
    if ost_a != 0:
        a += (n - ost_a)

    b -= (b % n)
    c = (b - a) // n + 1
    
    return (a + b) * c // 2
    


def get_result_3(a, b):
    c = 0
    
    c += get_number(a, b, 3)
    c += get_number(a, b, 5)
    
    c -= get_number(a, b, 15)
    
    return c


start = perf_counter()
print(get_result_3(1, 1_000 - 1))
end = perf_counter()

print(f'{end - start:.5f}')
```

### Оптимальное решение через компактную формулу

В этом блоке реализована максимально эффективная версия:

1. **func(n, k)** — вычисляет сумму всех чисел меньше **n**, кратных **k**, с помощью формулы суммы арифметической прогрессии
2. **get_result_4(n)** — использует принцип включений-исключений:  
   - Суммируем кратные 3 и 5
   - Вычитаем кратные 15, чтобы не считать их дважды
3. Этот метод не использует циклы, поэтому выполняется практически мгновенно, даже для больших значений **n**

```python
def func(n, k):
    m = (n - 1) // k
    return k * m * (m + 1) // 2

def get_result_4(n):
    return func(n, 3) + func(n, 5) - func(n, 15)


start = perf_counter()
print(get_result_4(1_000))
end = perf_counter()

print(f'{end - start:.5f}')
```

### Подготовка к тестированию на разных диапазонах

В этом блоке создаём список различных значений **n**, для которых будем измерять скорость работы всех четырёх методов

Также создаются пустые списки **results_1, results_2, results_3, results_4**, куда будем сохранять время выполнения каждого подхода 

Эти данные затем используются для построения графика сравнения производительности

```python
test_numbers = [1_000, 5_000, 10_000, 50_000, 100_000, 500_000, 1_000_000]

results_1 = []
results_2 = []
results_3 = []
results_4 = []
```

### Тестирование скорости выполнения всех методов

Проверка производительности всех четырёх решений на разных диапазонах **n**:

1. Для каждого значения из списка **test_numbers** запускаем каждый метод 50 раз, чтобы получить среднее время выполнения 
2. **results_1** — простой перебор через цикл `for` 
3. **results_2** — решение через функцию с циклом
4. **results_3** — математическое решение через арифметическую прогрессию
5. **results_4** — оптимальная компактная формула
6. После измерений выводим среднее время для каждого метода, чтобы сравнить их эффективность 

Эти данные позволяют построить график и наглядно увидеть разницу в скорости между подходами

```python
for test_number in test_numbers:
    rs = []

    for _ in range(50):
        start = perf_counter()
        
        c = 0
        for i in range(1, test_number):
            if i % 3 == 0 or i % 5 == 0:
                c += i
                
        end = perf_counter()
        rs.append(end - start)
        
    results_1.append(mean(rs))

for i, t in enumerate(test_numbers):
    print(t, f'{results_1[i]:0.10f}')
```

```python
for test_number in test_numbers:
    rs = []
    for _ in range(50):
        start = perf_counter()
        
        get_result_2(test_number)
        
        end = perf_counter()
        rs.append(end - start)
        
    results_2.append(mean(rs))

for i, t in enumerate(test_numbers):
    print(t, f'{results_2[i]:0.10f}')
```

```python
for test_number in test_numbers:
    rs = []
    for _ in range(50):
        start = perf_counter()
        
        get_result_3(1, test_number - 1)
        
        end = perf_counter()
        rs.append(end - start)
        
    results_3.append(mean(rs))

for i, t in enumerate(test_numbers):
    print(t, f'{results_3[i]:0.10f}')
```

```python
for test_number in test_numbers:
    rs = []
    for _ in range(50):
        start = perf_counter()
        
        get_result_4(test_number)
        
        end = perf_counter()
        rs.append(end - start)
        
    results_4.append(mean(rs))

for i, t in enumerate(test_numbers):
    print(t, f'{results_4[i]:0.10f}')
```

### Визуализация сравнения алгоритмов


Выводы по графику:

1. Методы с циклами (Algorithm 1 и 2) растут линейно с увеличением **n** и значительно медленнее на больших числах  
2. Математические методы (Algorithm 3 и особенно Algorithm 4) практически не зависят от размера диапазона и работают почти мгновенно 
3. Наибольшую эффективность показывает **Algorithm 4**, что делает его идеальным для больших значений **n**

```python
plt.figure(figsize=(16, 6))

plt.plot(test_numbers, results_1, marker='o', label='Algorithm 1')
plt.plot(test_numbers, results_2, marker='o', label='Algorithm 2')
plt.plot(test_numbers, results_3, marker='o', label='Algorithm 3')
plt.plot(test_numbers, results_4, marker='o', label='Algorithm 4')

plt.title('Сравнение алгоритмов')
plt.xlabel('Количество тестов')
plt.ylabel('Результаты')


plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

### Визуализация сравнения алгоритмов 3 и 4

Математические методы (Algorithm 3 и особенно Algorithm 4) работают практически мгновенно независимо от входных данных, при этом **Algorithm 4** показывает наибольшую эффективность

```python
plt.figure(figsize=(16, 6))


plt.plot(test_numbers, results_3, marker='o', label='Algorithm 3')
plt.plot(test_numbers, results_4, marker='o', label='Algorithm 4')

plt.title('Сравнение алгоритмов')
plt.xlabel('Количество тестов')
plt.ylabel('Результаты')


plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
