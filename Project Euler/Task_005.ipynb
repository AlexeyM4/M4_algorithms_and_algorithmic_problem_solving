## Используемые библиотеки

## 1) Через разложение чисел на простые множители

В этом подходе задача сводится к поиску **наименьшего общего кратного** всех чисел от `1` до `20`.

Идея заключается в том, чтобы разложить каждое число на простые множители и сохранить для каждого множителя **максимальную степень**, с которой он встречается.

Например:

```text
8 = 2³
12 = 2² · 3
```

Чтобы итоговое число делилось и на `8`, и на `12`, в нём должна присутствовать степень `2³`, потому что это максимальная степень двойки среди этих чисел.

### Алгоритм

1. Функция `get_divs(n)` раскладывает число `n` на простые множители.
2. Для каждого числа от `1` до `20` вычисляется его разложение.
3. В словаре `d` для каждого простого множителя хранится максимальная степень.
4. После обработки всех чисел вычисляется произведение:
   ```python
   prod(i**d[i] for i in d)
   ```
5. Полученное значение является искомым ответом.

### Особенности

- Решение строит ответ напрямую через простые множители.
- Словарь `d` хранит только максимальные степени, необходимые для делимости на все числа диапазона.
- Функция `prod()` делает перемножение множителей более читаемым.
- Перебор всех возможных кандидатов не используется.
- Время выполнения измеряется с помощью `perf_counter()`.

### Вывод

Разложение чисел на простые множители позволяет построить наименьшее общее кратное напрямую. Это делает решение достаточно быстрым и наглядно показывает математическую структуру задачи.

````python
def get_divs(n):
    r = {}
    div = 2
    while n > 1:
        if n % div == 0:
            r[div] = r.get(div, 0) + 1
            n //= div
        else:
            div += 1
    return r
        

start = perf_counter()
d = {}
for i in range(1, 20 + 1):
    t = get_divs(i)
    for j in t:
        d[j] = max(d.get(j, 0), t[j])

print(prod(i**d[i] for i in d))
end = perf_counter()

print(f'{end - start:.5f}')
````

## 2) Выделение основного алгоритма в отдельную функцию

В данной реализации логика решения полностью совпадает с предыдущим подходом.

Основное отличие заключается в том, что весь алгоритм вынесен в отдельную функцию `get_result_2()`, которая принимает верхнюю границу диапазона в качестве параметра.

Это делает решение более универсальным и позволяет легко использовать его для любых значений `n`, а не только для диапазона от `1` до `20`.

### Особенности

- Логика алгоритма полностью совпадает с предыдущей реализацией.
- Основной код выделен в отдельную функцию, что повышает читаемость и удобство повторного использования.
- Решение становится универсальным и подходит для любого значения `n`.
- В Python код, выполняемый внутри функции, зачастую работает немного быстрее благодаря более быстрому доступу к локальным переменным.

### Вывод

Выделение алгоритма в отдельную функцию делает решение более универсальным, удобным для повторного использования и последующего сравнения с другими подходами.

````python
def get_divs(n):
    r = {}
    div = 2
    while n > 1:
        if n % div == 0:
            r[div] = r.get(div, 0) + 1
            n //= div
        else:
            div += 1
    return r
        
def get_result_2(n):
    d = {}
    for i in range(1, n + 1):
        t = get_divs(i)
        for j in t:
            d[j] = max(d.get(j, 0), t[j])
            
    return prod(i**d[i] for i in d)


start = perf_counter()
print(get_result_2(20))
end = perf_counter()

print(f'{end - start:.5f}')
````

## 3) Последовательный поиск наименьшего общего кратного

В отличие от предыдущих решений, здесь не используется разложение чисел на простые множители.

Алгоритм начинает работу с максимального числа диапазона и последовательно строит **наименьшее общее кратное** для всех чисел от `20` до `1`.

На каждой итерации текущее значение увеличивается на свою первоначальную величину до тех пор, пока не станет кратным очередному числу.

### Алгоритм

1. Начальное значение `s` принимается равным `mx`.
2. Последовательно перебираются числа от `mx` до `1`.
3. Для каждого числа сохраняется текущее значение `s` в переменную `tmp`.
4. Пока `s` не делится на текущее число без остатка, к нему прибавляется `tmp`.
5. После обработки всех чисел значение `s` является искомым наименьшим общим кратным.

### Особенности

- Не требует разложения чисел на простые множители.
- Использует последовательное построение наименьшего общего кратного.
- Реализация достаточно компактна и легко читается.
- На каждой итерации выполняются только операции сложения и проверки делимости.
- Подходит для небольших значений `n`, однако при увеличении диапазона работает значительно медленнее математического метода.

### Вывод

Данный подход позволяет найти ответ без использования дополнительных математических преобразований. Несмотря на простоту реализации, по эффективности он уступает решению через разложение на простые множители, так как требует значительно большего количества операций.

````python
def get_result_3(mx):
    s = mx
    
    for i in range(mx, 0, -1):
        tmp = s
        while s % i != 0:
            s += tmp
    return s
    
    
start = perf_counter()
print(get_result_3(20))
end = perf_counter()

print(f'{end - start:.5f}')
````

## 4) Через НОД и НОК

В этом подходе задача решается через последовательное вычисление **наименьшего общего кратного** для чисел от `1` до `20`.

Используется важная связь между НОД и НОК двух чисел:

```text
НОК(a, b) = a · b / НОД(a, b)
```

Сначала вычисляется НОК для первых двух чисел, затем результат объединяется со следующим числом, потом со следующим и так далее.

### Алгоритм

1. Функция `get_NOD(a, b)` вычисляет наибольший общий делитель двух чисел с помощью алгоритма Евклида.
2. Функция `get_NOK(a, b)` вычисляет наименьшее общее кратное двух чисел через НОД.
3. Функция `reduce()` последовательно применяет `get_NOK` ко всем числам от `1` до `n`.
4. В результате получается число, которое делится на все числа диапазона.

### Особенности

- Решение напрямую использует связь между НОД и НОК.
- Алгоритм Евклида позволяет эффективно находить НОД.
- `reduce()` удобно применяет операцию НОК ко всему диапазону чисел.
- Код получается компактным и хорошо масштабируется для других значений `n`.
- Это один из самых аккуратных и математически естественных способов решения задачи.

### Вывод

Использование НОД и НОК позволяет получить короткое и эффективное решение. Такой подход хорошо подходит для задач, где необходимо найти число, кратное сразу нескольким значениям.

````python
def get_NOD(a, b):
    while a > 0 and b > 0:
        if a > b:
            a %= b
        else:
            b %= a
    return max(a, b)

def get_NOK(a, b):
    return (a * b) // get_NOD(a, b)
    
def get_result_4(n):
    return reduce(get_NOK, range(1, n + 1))

start = perf_counter()
print(get_result_4(20))
end = perf_counter()

print(f'{end - start:.5f}')
````

## 5) Использование встроенной функции `gcd()`

В предыдущем решении функции вычисления НОД и НОК были реализованы самостоятельно.

В данной реализации вычисление НОД полностью передаётся встроенной функции `gcd()` из модуля `math`, а вычисление НОК выполняется непосредственно внутри функции `reduce()`.

В результате код становится значительно компактнее, сохраняя ту же самую идею последовательного вычисления наименьшего общего кратного.

### Алгоритм

1. Последовательно перебираются числа от `1` до `n`.
2. Функция `reduce()` объединяет их, последовательно вычисляя НОК двух чисел.
3. Для вычисления НОК используется формула:
   ```text
   НОК(a, b) = a · b / НОД(a, b)
   ```
4. НОД вычисляется встроенной функцией `gcd()`.
5. После обработки всего диапазона возвращается искомое число.

### Особенности

- Используется встроенная функция `gcd()` из стандартной библиотеки Python.
- Отсутствует необходимость самостоятельно реализовывать алгоритм Евклида.
- Код становится значительно компактнее и легче читается.
- Основная логика полностью совпадает с предыдущим решением.
- Встроенные функции Python обычно работают быстрее аналогичных реализаций на чистом Python.

### Вывод

Использование стандартной библиотеки позволяет существенно сократить объём кода без изменения алгоритма. Такой вариант является наиболее лаконичным и хорошо демонстрирует преимущества использования готовых инструментов языка.

````python
def get_result_5(n):
    return reduce(lambda a, b: (a * b) // gcd(a, b), range(1, n + 1))

start = perf_counter()
print(get_result_5(20))
end = perf_counter()

print(f'{end - start:.5f}')
````
