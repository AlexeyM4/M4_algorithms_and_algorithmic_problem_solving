### Задача 3

Простые делители числа 13195 - это 5, 7, 13 и 29.

Каков самый большой делитель числа 600851475143, являющийся простым числом?

Ответ: 6857

## Подключение библиотек

Подключение вспомогательных инструментов, которые пригодятся для решения задачи и анализа производительности.

- `perf_counter` из модуля `time` — используется для точного измерения времени выполнения программы. Это позволит сравнить скорость разных подходов к решению задачи.

- `isprime` из библиотеки `sympy` — функция для проверки, является ли число простым. Она пригодится при поиске простых делителей числа.

- `mean` из модуля `statistics` — используется для вычисления среднего значения.

```python
from time import perf_counter
from sympy import isprime
from statistics import mean
```

## Первый способ решения

В данном варианте реализована собственная функция проверки числа на простоту и выполнен перебор делителей числа `600851475143`.

### Проверка числа на простоту

Функция `is_prime_1` работает следующим образом:

- Если число меньше либо равно 1 — оно не является простым.
- Перебор возможных делителей выполняется от `2` до `√num`.
- Если находится делитель, число составное.
- Если делителей нет — число простое.

Проверка до квадратного корня достаточна, так как делители у числа всегда идут парами.

---

### Поиск наибольшего простого делителя

Алгоритм перебирает числа от `1` до `√n`.  
Если число `i` является делителем `n`, вычисляется парный делитель `j = n // i`.

Сначала проверяется больший делитель `j`.  
Если он простой, поиск завершается, так как найден максимальный простой делитель.

Если `j` не является простым, выполняется проверка числа `i`.

---

### Измерение времени

С помощью `perf_counter()` определяется время выполнения алгоритма для оценки его производительности.

```python
def is_prime_1(num: int) -> bool:
    if num <= 1:
        return False
    
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
        
    return True


start = perf_counter()

n = 600851475143
md = 0
for i in range(1, int(n ** 0.5) + 1):
    if n % i == 0:
        j = n // i
        if is_prime_1(j):
            md = j
            break
        if is_prime_1(i):
            md = i
print(md)
        
end = perf_counter()

print(f'{end - start:.5f}')
```

## Второй вариант решения

В данном случае используется тот же алгоритм поиска наибольшего простого делителя, однако основная логика вынесена в отдельную функцию `get_result_2`.

Структура решения не изменилась:

- выполняется перебор делителей до `√n`;
- при нахождении делителя проверяется парный делитель;
- используется функция проверки числа на простоту;
- при обнаружении максимального простого делителя выполнение завершается.

Главное отличие — логика вычислений оформлена в виде функции.  

В Python выполнение кода внутри функции обычно работает немного быстрее, чем аналогичный код на верхнем уровне скрипта, поскольку локальные переменные обрабатываются эффективнее глобальных.

```python
def is_prime_2(num: int) -> bool:
    if num <= 1:
        return False
    
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


def get_result_2(n):
    n = 600851475143
    md = 0
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            j = n // i
            if is_prime_2(j):
                return j
            
            if is_prime_2(i):
                md = i
    return md


start = perf_counter()
print(get_result_2(600851475143))
end = perf_counter()

print(f'{end - start:.5f}')
```

## Третий вариант решения

В третьем варианте реализована более оптимизированная проверка числа на простоту и улучшен поиск наибольшего простого делителя.

### Проверка числа на простоту

Функция `is_prime_3` работает быстрее за счёт сокращения количества проверок:

- Если число ≤ 1 — оно не является простым.
- Число 2 считается простым.
- Чётные числа сразу исключаются.
- Далее проверяются только нечётные делители.

Перебор делителей выполняется от 3 до √num с шагом 2, так как чётные числа уже не рассматриваются.

---

### Поиск результата

Функция `get_result_3` выполняет поиск наибольшего простого делителя:

- Перебираются возможные делители числа `n` до √n.
- Если найден делитель:
  - Сначала проверяется парный делитель `j = n // i`, так как он потенциально может быть максимальным.
  - Если `j` является простым — функция сразу возвращает результат.
  - Если нет — проверяется делитель `i`, который сохраняется как текущий кандидат.

Такой подход позволяет быстрее завершить вычисления при нахождении максимального простого делителя.

---

### Измерение времени выполнения

Как и в предыдущих вариантах, используется функция `perf_counter()` для оценки производительности алгоритма.

Этот вариант является наиболее оптимизированным среди представленных за счёт:
- уменьшения количества проверок делителей;
- исключения чётных чисел из перебора;
- раннего завершения поиска при нахождении результата.

```python
def is_prime_3(num: int) -> bool:
    if num <= 1:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    
    for i in range(3, int(num ** 0.5) + 1, 2):
        if num % i == 0:
            return False
    return True


def get_result_3(n):
    n = 600851475143
    md = 0
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            j = n // i
            if is_prime_3(j):
                return j
            
            if is_prime_3(i):
                md = i
    return md



start = perf_counter()
print(get_result_3(600851475143))
end = perf_counter()

print(f'{end - start:.5f}')
```

### Четвёртый вариант решения

В данном варианте для проверки простоты числа используется функция `isprime` из библиотеки `sympy`.

Основное преимущество этого подхода — высокая скорость и надёжность проверки простоты, так как функция уже оптимизирована и реализована на уровне математической библиотеки.

Однако стоит учитывать, что библиотека `sympy` устанавливается отдельно через `pip install sympy`.

Использование готовой функции проверки простоты позволяет сократить объём собственного кода и упростить реализацию алгоритма поиска наибольшего простого делителя.

```python
def get_result_4(n):
    n = 600851475143
    md = 0
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            j = n // i
            if isprime(j):
                return j
            
            if isprime(i):
                md = i
    return md


start = perf_counter()
print(get_result_4(600851475143))
end = perf_counter()

print(f'{end - start:.5f}')
```

### Анализ времени выполнения алгоритмов

Создаётся список `results`, в который будут сохраняться средние времена выполнения каждого алгоритма.  
Переменная `n` задаёт количество повторений измерения времени для повышения точности оценки.

```python
results = []

t = 20
```

Измеряется среднее время выполнения первого алгоритма.

- Код запускается `t` раз для уменьшения влияния случайных отклонений.
- Для каждого запуска фиксируется время выполнения поиска наибольшего простого делителя.
- Результаты измерений сохраняются в список `rs`.
- В список `results` добавляется среднее значение времени.

```python
rs = []
for _ in range(t):
    start = perf_counter()

    n = 600851475143
    md = 0
    for i in range(1, int(n ** 0.5) + 1):
        if n % i == 0:
            j = n // i
            if is_prime_1(j):
                md = j
                break
            if is_prime_1(i):
                md = i

    end = perf_counter()
    rs.append(end - start)
    
results.append(mean(rs))
```

Измеряется время работы остальных алгоритмов (`get_result_2`, `get_result_3`, `get_result_4`).

- Для каждой функции выполняется `t` запусков.
- Вычисляется среднее время выполнения функции.
- Это позволяет корректно сравнить производительность разных реализаций.

```python
funcs = (get_result_2, get_result_3, get_result_4)

for func in funcs:
    rs = []
    for _ in range(t):
        start = perf_counter()

        func(600851475143)

        end = perf_counter()
        rs.append(end - start)
    
    results.append(mean(rs))
```

Выводятся средние времена выполнения каждого алгоритма.  
Числа округляются до 5 знаков после запятой.

```python
for i, res in enumerate(results, start=1):
    print(f'Алгоритм {i}: {res:.5f}')
```
