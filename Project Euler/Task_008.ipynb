### Задача 8

Наибольшее произведение четырёх последовательных цифр в приведённом ниже 1000-значном числе равно

$$
9 \cdot 9 \cdot 8 \cdot 9 = 5832

## 1) Полный перебор последовательностей

В данном решении последовательно рассматриваются все возможные группы из **13 соседних цифр** исходного числа.

Для каждой такой группы вычисляется произведение всех цифр, после чего найденное значение сравнивается с текущим максимумом.

### Алгоритм

1. Исходное число преобразуется в строку.
2. Последовательно перебираются все возможные подстроки длиной `13`.
3. Для каждой подстроки вычисляется произведение её цифр с помощью функции `reduce()`.
4. Если найденное произведение больше текущего максимума, значение обновляется.
5. После обработки всех последовательностей возвращается наибольшее произведение.

### Особенности

- Используется полный перебор всех последовательностей длиной `13`.
- Произведение каждой последовательности вычисляется независимо от остальных.
- Для вычисления произведения применяется функция `reduce()`.
- При переходе к следующей последовательности произведение вычисляется заново, что приводит к повторным вычислениям.

### Вывод

Полный перебор является наиболее простым способом решения задачи. Несмотря на наличие повторных вычислений, он остаётся понятным и позволяет получить правильный ответ без использования дополнительных оптимизаций.

```python
def get_result_1(n: int) -> int:
    sn = str(n)
    mx = 0
    for i in range(len(sn) - 12):
        mx = max(mx, reduce(lambda a, b: int(a) * int(b), sn[i:i + 13]))
    return mx
   

start = perf_counter()
print(get_result_1(num))
end = perf_counter()

print(f'{end - start:.5f}')
```

## 2) Исключение последовательностей с нулём

В этой реализации используется та же идея полного перебора последовательностей длиной `13`, однако перед началом вычислений исходное число разделяется по цифре `0`.

Любая последовательность, содержащая ноль, имеет произведение, равное `0`, поэтому такие участки можно сразу исключить из рассмотрения.

### Алгоритм

1. Исходное число преобразуется в строку.
2. Строка разделяется по символу `0`.
3. Сохраняются только участки длиной не менее `13`, поскольку более короткие фрагменты не могут содержать нужную последовательность.
4. Для каждого подходящего фрагмента перебираются все подстроки длиной `13`.
5. Для каждой подстроки вычисляется произведение цифр.
6. Возвращается максимальное найденное значение.

### Особенности

- Последовательности, содержащие ноль, заранее исключаются.
- Участки длиной меньше `13` также не рассматриваются.
- Количество вычислений уменьшается по сравнению с первым решением.
- Произведение каждой подходящей последовательности по-прежнему вычисляется заново.
- Общая идея алгоритма остаётся прежней.

### Вывод

Предварительное исключение участков с нулями позволяет избежать заведомо бесполезных вычислений. Решение становится эффективнее первого подхода, сохраняя при этом простую и понятную структуру.

```python
def get_result_2(n: int) -> int:
    sn = [i for i in str(n).split('0') if len(i) >= 13]

    mx = 0
    for st in sn:
        for i in range(len(st) - 12):
            mx = max(mx, reduce(lambda a, b: int(a) * int(b), st[i:i + 13]))
    return mx

        
start = perf_counter()
print(get_result_2(num))
end = perf_counter()

print(f'{end - start:.5f}')
```

## 3) Метод скользящего окна

В данном решении используется техника **скользящего окна** (*Sliding Window*), позволяющая избежать повторного вычисления произведения для каждой последовательности.

Как и в предыдущем подходе, исходное число предварительно разделяется по цифре `0`, после чего рассматриваются только фрагменты длиной не менее `13`.

Для первой последовательности произведение вычисляется полностью. При переходе к следующему окну выполняется всего две операции:

- произведение делится на цифру, покидающую окно;
- умножается на новую цифру, входящую в окно.

Таким образом, нет необходимости заново перемножать все тринадцать цифр.

### Алгоритм

1. Исходное число разделяется по цифре `0`.
2. Рассматриваются только фрагменты длиной не менее `13`.
3. Для первых тринадцати цифр вычисляется начальное произведение.
4. Далее окно последовательно сдвигается на одну позицию.
5. При каждом сдвиге произведение обновляется делением на выходящую цифру и умножением на новую.
6. После обработки всех окон возвращается максимальное произведение.

### Особенности

- Используется метод **скользящего окна**.
- Последовательности с нулями исключаются заранее.
- Для каждого нового окна выполняются только две арифметические операции вместо полного пересчёта произведения.
- Встроенная функция `prod()` используется только один раз для вычисления начального произведения.
- Решение значительно эффективнее предыдущих подходов.

### Вывод

Метод скользящего окна позволяет существенно сократить количество вычислений за счёт повторного использования уже найденного произведения. Благодаря этому данный подход является наиболее эффективным среди рассмотренных решений.

```python
from math import prod

def get_result_3(n: int) -> int:
    sn = (list(map(int, i)) for i in str(n).split('0') if len(i) >= 13)

    mx = 0
    for st in sn:
        pr = prod(st[:13])
        mx = max(mx, pr)
        for i in range(13, len(st)):
            pr = (pr // st[i - 13]) * st[i]
            mx = max(mx, pr)
    return mx


start = perf_counter()
print(get_result_3(num))
end = perf_counter()

print(f'{end - start:.5f}')
```

## Сравнение времени выполнения

Для сравнения производительности используется функция `timeit()`, предназначенная специально для измерения времени выполнения небольших участков кода.

Каждый алгоритм запускается **1000 раз**, после чего общее время делится на количество запусков. Таким образом получается **среднее время выполнения одного запуска**.

```python
from timeit import timeit

funcs = (
    get_result_1,
    get_result_2,
    get_result_3,
)

for i, func in enumerate(funcs, start=1):
    tm = timeit(lambda: func(num), number=1000) / 1000
    print(f'Алгоритм {i}: {tm:.5f}')
```

## Заключение

В ходе решения задачи были рассмотрены три различных подхода к поиску наибольшего произведения тринадцати последовательных цифр.

Первый подход основан на полном переборе всех возможных последовательностей. Несмотря на простоту реализации, он выполняет большое количество повторных вычислений и оказался самым медленным.

Во втором решении удалось повысить производительность за счёт предварительного исключения последовательностей, содержащих цифру `0`. Это позволило существенно сократить количество рассматриваемых вариантов и ускорить выполнение алгоритма.

Наиболее эффективным оказалось третье решение, использующее метод **скользящего окна** (*Sliding Window*). Вместо повторного вычисления произведения для каждой новой последовательности оно обновляет уже найденное значение, выполняя лишь деление и умножение. Благодаря этому время выполнения оказалось почти в **18 раз меньше**, чем у первого решения.

Среднее время выполнения алгоритмов:

- **Алгоритм 1:** `0.00176` с
- **Алгоритм 2:** `0.00051` с
- **Алгоритм 3:** `0.00010` с

Данный пример наглядно показывает, что применение подходящей алгоритмической идеи позволяет значительно повысить производительность программы.
